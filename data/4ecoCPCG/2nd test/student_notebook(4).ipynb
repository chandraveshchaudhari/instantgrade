### 1. Get value with default

**Return:** The value for `key` in dictionary `d` if present, otherwise return the provided `default` value.

```python
def get_with_default(d: dict, key, default=None):
    return d.get(key, default)

get_with_default({'a': 1, 'b': 2, 'c': 3}, 'b', 0)
```

### 2. Get list of keys

**Return:** A list of keys from the dictionary (order not required).

```python
def keys_list(d: dict) -> list:
    return list(d.keys())

keys_list({'a': 1, 'b': 2, 'c': 3})
```

### 3. Merge two dictionaries

**Return:** A new dictionary containing keys from both `a` and `b` with values from `b` overriding `a` where keys overlap.

```python
def merge_dicts(a: dict, b: dict) -> dict:
    merged = a.copy()
    merged.update(b)
    return merged 
```

### 4. Check key exists

**Return:** `True` if `key` exists in `d`, otherwise `False`.

```python
def has_key(d: dict, key) -> bool:
    return key in d
```

### 5. Unique elements from list (set creation)

**Return:** A set containing unique elements from the given list.

```python
def unique_set(lst: list) -> set:
    return set(lst)
```

### 6. Count occurrences of items

**Return:** A dictionary mapping each item to its count in the list.

```python
def count_occurrences(lst: list) -> dict:
    counts = {}
    for item in lst: 
        counts[item] = counts.get(item, 0) + 1
    return counts 
```

### 7. Intersection of two sets

**Return:** A set containing elements common to both input sets.

```python
def set_intersection(a: set, b: set) -> set:
    return a.intersection(b) 
```

### 8. Remove duplicates preserving order

**Return:** A list containing elements from the input list with duplicates removed but original order preserved. (Use dict or ordered approach.)

```python
def remove_duplicates_preserve_order(lst: list) -> list:
    return list(dict.fromkeys(lst))
```

### 9. Invert a dictionary

**Return:** A new dictionary where original values become keys and values are lists of original keys that had that value.

You provided this sample; keep the behaviour (use lists to collect multiple keys for the same value).

```python
def invert_dict(d: dict) -> dict:
    inverted = {}
    for key, value in d.items():
        if value not in inverted:
            inverted[value] = []
            inverted[value].append(key)
            return inverted
            
```
