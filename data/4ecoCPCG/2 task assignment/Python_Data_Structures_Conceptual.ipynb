# 📘 Python Data Structures – Conceptual Explanation

## 🧩 Introduction
Python is a high-level programming language widely used for data handling and problem-solving.

## 🔑 Key Concepts
Data structures are containers used to store data efficiently.

### 1️⃣ List
Ordered and mutable collection.
```python
prices = [100, 120, 150]
```

### 2️⃣ Tuple
Ordered but immutable.
```python
coordinates = (12.5, 77.6)
```

### 3️⃣ Dictionary
Key–value pairs.
```python
student = {'name': 'Amit', 'marks': 85}
```

### 4️⃣ Set
Unordered and unique.
```python
countries = {'India', 'USA', 'UK'}
```

## 🏢 Real-Life Use Case
Used in economics and business data handling.

## ✅ Advantages
Easy, flexible, beginner-friendly.

## ⚠️ Limitations
Memory usage and speed issues.

## 📝 Summary
Python data structures form the foundation of efficient programming.
