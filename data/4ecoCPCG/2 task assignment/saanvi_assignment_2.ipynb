## Introduction

data access, modification, and interpretation easier.

In **economics, business analytics, and finance**, Python data structures are widely used to

## Key Concepts

### What are Data Structures?
- Methods of organizing and storing data
- Enable efficient access and modification
- Form the foundation of programming and data analysis

### Main Python Data Structures
1. **List** – Ordered and mutable collection  
2. **Tuple** – Ordered but immutable collection  
3. **Set** – Unordered collection of unique elements  
4. **Dictionary** – Stores data in key–value pairs  

---

## Explanation of Python Data Structures

### 1. List
- Stores multiple values in a single variable
- Allows duplicate elements
- Elements can be changed after creation

### 2. Tuple
- Similar to lists but **cannot be modified**
- Faster and more secure than lists
- Used for fixed data

### 3. Set
- Stores **only unique values**
- Automatically removes duplicates
- Does not follow a specific order

### 4. Dictionary
- Stores data as **key–value pairs**
- Keys must be unique
- Useful for labeled or structured data


```python
# List: monthly inflation rates
inflation_rates = [5.1, 5.3, 5.6, 5.4]

# Tuple: fixed base year
base_year = ("apple","banana")

# Set: unique sectors in an economy
sectors = {"Agriculture", "Industry", "Services"}

# Dictionary: GDP by sector
gdp = {
    "Agriculture": 18,
    "Industry": 29,
    "Services": 53
}
```

```python

# -------------------------
# Stores data in key–value pairs

student = {
    "name": "Asha",
    "age": 20,
    "course": "Economics"
}

student["age"] = 21          # updating a value
student["college"] = "ABC"   # adding a new key-value pair
```

##Real-Life or Business Use Case

Dictionaries: Storing GDP, population, or budget data by country
