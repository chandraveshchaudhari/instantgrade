### 1. Check if a number is positive, negative, or zero
**Return:** A string such as `'Positive'`, `'Negative'`, or `'Zero'`

```python
def check_number(n: int) -> str:
    if n > 0:
      return 'Positive'
    elif n < 0:
      return 'Negative'
    else:
      return 'Zero'

check_number(5)
```

### 2. Return a list of numbers from 1 to n (where n can be anything).
Example if n = 5 then
**Return:** `[1, 2, 3, 4, 5]`

```python
def numbers_1_to_n(n) -> list[int]:
    return list(range(1, n + 1))
numbers_1_to_n(5)
```

### 3. Return only even numbers from 1 to n (where n can be anything).
Example if n = 11 then
**Return:** `[2, 4, 6, 8, 10]`

```python
def even_numbers_1_to_n(n) -> list[int]:
    return [i for i in range(2, n + 1, 2)]
even_numbers_1_to_n(11)
```

### 4. Return the square of a given number
**Example:** For input `4`, return `16`

```python
def square_number(n: int) -> int:
    return n*n
square_number(4)
```

### 5. Return `'Access Granted'` if the password is `'python'`, else `'Access Denied'`
**Example:** Input `'java'` â†’ `'Access Denied'`; Input `'python'` â†’ `'Access Granted'`

```python
def check_password(password: str) -> str:
    if password == "python":
        return "Access Granted"
    else:
        return "Access Denied"
user_input = input()
check_password(user_input)
```

### 7. Return `'Even'` if a number is even, else `'Odd'`
**Example:** Input `4` â†’ `'Even'`

```python
def even_or_odd(n: int) -> str:
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(4))
```

### 9. Return a list of squares for numbers 1 to n
if n is 5
**Return:** `[1, 4, 9, 16, 25]`

```python
def squares_1_to_n(n: int) -> list[int]:
    squares = []
    for i in range(1, n + 1):
        squares.append(i * i)
    return squares



print(squares_1_to_n(5))
```

### 10. Count how many times a letter appears in a word  
**Example:** Input `('banana', 'a')`
**Return:** `3`

```python
def count_letter(word: str, letter: str) -> int:
    return word.count(letter)



print(count_letter("banana", "a"))
```

### 11. Return the average of three numbers  
**Example:** Input `(4, 8, 10)` â†’ `7.3333...`  
**Return:** The arithmetic mean of the three numbers.

```python
def average_of_three(a: float, b: float, c: float) -> float:
    return (a + b + c) / 3


print(average_of_three(4, 8, 10))
```
