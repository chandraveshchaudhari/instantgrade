# **Introduction**


Python is a high level, interpreted programming language , it was created by Guido van Rossum and first released in 1991. It is used for data scraping, analysis, modeling, and visualization

##History

###Late 1980s – 1990s:
*   Python was created as a general purpose language by Guido van Rossum.
*   economists primarily used specialized tools like Fortran or MATLAB for computation.
###Early 2000s:
*   The release of NumPy and SciPy provided the help required for scientific computing
###2008 - 2009:
*   A major update was released to make the language cleaner and more modern.
*   The invention of pandas made handling structured data like Excel spreadsheets and SQL tables significantly easier,which boosted productivity and improved the data driven economy
###2014 – Present:
*  The launch of QuantEcon marked a major shift, providing high quality, open source lectures that brought Python into PhD curricula worldwide.
*   Major institutions like the Federal Reserve, IMF, and World Bank have transitioned to Python for  forecasting and real time policy simulation.

# **Advantages: Python vs. Others**
* Versatility: It is a general purpose language. You can use it for data analysis, web development, and task automation, but tools like R or Stata are specialized only for statistics.
* Readability: The syntax mimics English, making it much easier to learn and maintain than "dense" languages like C++ or Java.
* Cost : It is completely free open-source and has the world's largest programming community, meaning help and documentation are always available.

# **Limitations**
* Speed: It is slower than  languages like C++,java , c# etc
* Memory Consumption: It uses more memory compared  to lower-level languages.
