### 1. Check if a number is positive, negative, or zero
**Return:** A string such as `'Positive'`, `'Negative'`, or `'Zero'`

```python
def check_number(n: int) -> str:
    if n > 0:
      return 'Positive'
    elif n < 0:
      return 'Negative'
    else:
      return 'Zero'

check_number(5)
```

### 2. Return a list of numbers from 1 to n (where n can be anything).
Example if n = 5 then
**Return:** `[1, 2, 3, 4, 5]`

```python
def numbers_1_to_n(n) -> list[int]: return list(range(1,n+1))
n=5
print(numbers_1_to_n(n))
    
  
```

### 3. Return only even numbers from 1 to n (where n can be anything).
Example if n = 11 then
**Return:** `[2, 4, 6, 8, 10]`

```python
def even_numbers_1_to_n(n) -> list[int]: return [i for i in range(1, n+1) if i % 2 == 0] 
n = 25
print(even_numbers_1_to_n(n))
    
```

### 4. Return the square of a given number
**Example:** For input `4`, return `16`

```python
def square_number(n: int) -> int: return n*n
n=4
print(square_number(n))
```

### 5. Return `'Access Granted'` if the password is `'python'`, else `'Access Denied'`
**Example:** Input `'java'` → `'Access Denied'`; Input `'python'` → `'Access Granted'`

```python
def check_password(password: str) -> str: return " Access Granted" if password== "python" else "Access Denied"
password=input("enter password")
print(check_password(password))
```

### 6. Return the larger of two numbers  
**Example:** Input `(4, 9)` → `9`  
**Return:** The larger number.

```python
def larger_number(a: int, b: int) -> int: return a if a > b else b
a = input("enter a number")
b= input("enter another number")
print(larger_number(a,b))
```

### 7. Return `'Even'` if a number is even, else `'Odd'`
**Example:** Input `4` → `'Even'`

```python
def even_or_odd(n: int) -> str: return "Number is Even" if n%2==0 else "Number is odd"
n=int(input("Enter a no."))
print(even_or_odd(n))
```

### 9. Return a list of squares for numbers 1 to n
if n is 5
**Return:** `[1, 4, 9, 16, 25]`

```python
def squares_1_to_n(n) -> list[int]: return[i*i for i in range(1, n+1)]
n=5
print(squares_1_to_n(n))
```

### 10. Count how many times a letter appears in a word  
**Example:** Input `('banana', 'a')`
**Return:** `3`

```python
def count_letter(word: str, letter: str) -> int: return word.count(letter)
word=input("enter a word")
letter=input("enter a letter in the word")
print(count_letter(word,letter))
```

### 11. Return the average of three numbers  
**Example:** Input `(4, 8, 10)` → `7.3333...`  
**Return:** The arithmetic mean of the three numbers.

```python
def average_of_three(a: float, b: float, c: float) -> float: return(a+b+c)/3
a=float(input("enter number a"))
b=float(input("enter number b"))
c=float(input("enter number c"))
print(average_of_three(a,b,c))
```
