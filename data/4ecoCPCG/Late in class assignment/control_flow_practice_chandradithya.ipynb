### 1. Check if a number is positive, negative, or zero

```python
def check_number(n: int) -> str:
    if n > 0:
        return 'Positive'
    elif n < 0:
        return 'Negative'
    else:
        return 'Zero'
```

### 2. Return numbers from 1 to n

```python
def numbers_1_to_n(n) -> list[int]:
    return list(range(1, n + 1))
```

### 3. Return even numbers from 1 to n

```python
def even_numbers_1_to_n(n) -> list[int]:
    return [i for i in range(2, n + 1, 2)]
```

### 4. Square of a number

```python
def square_number(n: int) -> int:
    return n * n
```

### 5. Password check

```python
def check_password(password: str) -> str:
    return 'Access Granted' if password == 'python' else 'Access Denied'
```

### 6. Larger of two numbers

```python
def larger_number(a: int, b: int) -> int:
    return a if a > b else b
```

### 7. Even or Odd

```python
def even_or_odd(n: int) -> str:
    return 'Even' if n % 2 == 0 else 'Odd'
```

### 8. Sum from 1 to n

```python
def sum_1_to_n(n: int) -> int:
    return sum(range(1, n + 1))
```

### 9. Squares from 1 to n

```python
def squares_1_to_n(n) -> list[int]:
    return [i * i for i in range(1, n + 1)]
```

### 10. Count letter occurrences

```python
def count_letter(word: str, letter: str) -> int:
    return word.count(letter)
```

### 11. Average of three numbers

```python
def average_of_three(a: float, b: float, c: float) -> float:
    return (a + b + c) / 3
```
