# What is Python?

## Introduction
Python is a high-level programming language known for its simplicity and readability.
It is widely used in education, business, data analysis, and web development.


## Key Concepts

- Python is an interpreted language
- It uses simple and readable syntax
- It supports multiple programming styles
- It has a large standard library


```python
x = 5
y = 10
x + y
```

## Real-Life or Business Use Case

Python is used by businesses to automate repetitive tasks, analyze market trends,
and manage large amounts of data efficiently.


```python
hours_worked = 40
hourly_wage = 500
salary = hours_worked * hourly_wage
salary
```

## Advantages

- Easy to learn and use
- Open-source and free
- Large community support
- Used across many industries


## Limitations

- Slower compared to some compiled languages
- Not ideal for mobile app development
- High memory usage in some cases


## Summary

Python is a versatile and beginner-friendly programming language.
Its simplicity and wide range of applications make it one of the most popular
languages today.
