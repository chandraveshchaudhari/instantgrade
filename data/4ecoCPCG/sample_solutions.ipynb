## Check if a number is positive, negative, or zero
**Return:** A string such as `'Positive'`, `'Negative'`, or `'Zero'`

```python
def check_number(n: int) -> str:
    if n > 0:
        return 'Positive'
    elif n < 0:
        return 'Negative'
    else:
        return 'Zero'
```

```python
assert check_number(5) == 'Positive'
assert check_number(-3) == 'Negative'
assert check_number(0) == 'Zero'
```

## Return a list of numbers from 1 to n
**Example:** If n = 5, return `[1, 2, 3, 4, 5]`

```python
def numbers_1_to_n(n: int) -> list[int]:
    return list(range(1, n + 1))
```

```python
assert numbers_1_to_n(5) == [1, 2, 3, 4, 5]
assert numbers_1_to_n(1) == [1]
assert numbers_1_to_n(0) == []
```

## Return only even numbers from 1 to n
**Example:** If n = 11, return `[2, 4, 6, 8, 10]`

```python
def even_numbers_1_to_n(n: int) -> list[int]:
    return [i for i in range(1, n + 1) if i % 2 == 0]
```

```python
assert even_numbers_1_to_n(11) == [2, 4, 6, 8, 10]
assert even_numbers_1_to_n(6) == [2, 4, 6]
assert even_numbers_1_to_n(1) == []
```

## Return the square of a given number
**Example:** For input `4`, return `16`

```python
def square_number(n: int) -> int:
    return n * n
```

```python
assert square_number(4) == 16
assert square_number(5) == 25
assert square_number(0) == 0
```

## Return 'Access Granted' if password is 'python', else 'Access Denied'
**Example:** Input `'java'` → `'Access Denied'`; Input `'python'` → `'Access Granted'`

```python
def check_password(password: str) -> str:
    if password == "python":
        return "Access Granted"
    else:
        return "Access Denied"
```

```python
assert check_password("python") == "Access Granted"
assert check_password("java") == "Access Denied"
assert check_password("PYTHON") == "Access Denied"
```

## Return the larger of two numbers
**Example:** Input `(4, 9)` → `9`

```python
def larger_number(a: int, b: int) -> int:
    if a > b:
        return a
    else:
        return b
```

```python
assert larger_number(4, 9) == 9
assert larger_number(10, 5) == 10
assert larger_number(7, 7) == 7
```

## Return 'Even' if a number is even, else 'Odd'
**Example:** Input `4` → `'Even'`

```python
def even_or_odd(n: int) -> str:
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
assert even_or_odd(4) == "Even"
assert even_or_odd(5) == "Odd"
assert even_or_odd(0) == "Even"
```
