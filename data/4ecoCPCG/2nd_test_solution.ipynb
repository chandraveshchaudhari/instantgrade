### 1. Get value with default

**Return:** The value for `key` in dictionary `d` if present, otherwise return the provided `default` value.

```python
def get_with_default(d: dict, key, default=None):
    return d.get(key, default)
```

```python
# Test get_with_default
assert get_with_default({'a': 1, 'b': 2, 'c': 3}, 'b', 0) == 2
assert get_with_default({'a': 1, 'b': 2}, 'x', 'default') == 'default'
assert get_with_default({}, 'a', None) is None
```

### 2. Get list of keys

**Return:** A list of keys from the dictionary (order not required).

```python
def keys_list(d: dict) -> list:
    return list(d.keys())
```

```python
# Test keys_list
result = keys_list({'a': 1, 'b': 2, 'c': 3})
assert set(result) == {'a', 'b', 'c'}
assert len(result) == 3
assert keys_list({}) == []
```

### 3. Merge two dictionaries

**Return:** A new dictionary containing keys from both `a` and `b` with values from `b` overriding `a` where keys overlap.

```python
def merge_dicts(a: dict, b: dict) -> dict:
    return {**a, **b}
```

```python
# Test merge_dicts
assert merge_dicts({'x': 1, 'y': 2}, {'y': 10, 'z': 3}) == {'x': 1, 'y': 10, 'z': 3}
assert merge_dicts({}, {'a': 1}) == {'a': 1}
assert merge_dicts({'a': 1}, {}) == {'a': 1}
```

### 4. Check key exists

**Return:** `True` if `key` exists in `d`, otherwise `False`.

```python
def has_key(d: dict, key) -> bool:
    return key in d
```

```python
# Test has_key
assert has_key({'a': 1, 'b': 2}, 'a') is True
assert has_key({'a': 1, 'b': 2}, 'c') is False
assert has_key({}, 'x') is False
```

### 5. Unique elements from list (set creation)

**Return:** A set containing unique elements from the given list.

```python
def unique_set(lst: list) -> set:
    return set(lst)
```

```python
# Test unique_set
assert unique_set([1, 2, 2, 3, 4, 4]) == {1, 2, 3, 4}
assert unique_set([]) == set()
assert unique_set([1, 1, 1]) == {1}
```

### 6. Count occurrences of items

**Return:** A dictionary mapping each item to its count in the list.

```python
def count_occurrences(lst: list) -> dict:
    counts = {}
    for item in lst:
        counts[item] = counts.get(item, 0) + 1
    return counts
```

```python
# Test count_occurrences
assert count_occurrences([1, 2, 2, 3, 3, 3]) == {1: 1, 2: 2, 3: 3}
assert count_occurrences([]) == {}
assert count_occurrences(['a', 'a', 'b']) == {'a': 2, 'b': 1}
```

### 7. Intersection of two sets

**Return:** A set containing elements common to both input sets.

```python
def set_intersection(a: set, b: set) -> set:
    return a & b
```

```python
# Test set_intersection
assert set_intersection({1, 2, 3}, {2, 3, 4}) == {2, 3}
assert set_intersection({1, 2}, {3, 4}) == set()
assert set_intersection({1, 2}, {1, 2}) == {1, 2}
```

### 8. Remove duplicates preserving order

**Return:** A list containing elements from the input list with duplicates removed but original order preserved. (Use dict or ordered approach.)

```python
def remove_duplicates_preserve_order(lst: list) -> list:
    return list(dict.fromkeys(lst))
```

```python
# Test remove_duplicates_preserve_order
assert remove_duplicates_preserve_order([1, 2, 2, 3, 1, 4]) == [1, 2, 3, 4]
assert remove_duplicates_preserve_order([]) == []
assert remove_duplicates_preserve_order([1]) == [1]
```

### 9. Invert a dictionary

**Return:** A new dictionary where original values become keys and values are lists of original keys that had that value.

```python
def invert_dict(d: dict) -> dict:
    inverted = {}
    for key, value in d.items():
        if value not in inverted:
            inverted[value] = []
        inverted[value].append(key)
    return inverted
```

```python
# Test invert_dict
result = invert_dict({'a': 1, 'b': 1, 'c': 2})
assert result == {1: ['a', 'b'], 2: ['c']}
assert invert_dict({}) == {}
assert invert_dict({'x': 5}) == {5: ['x']}
```

### 10. Find values that have multiple keys

**Return:** A dictionary mapping each value that appears for more than one key to the list of keys that map to it. (Exclude values that are unique.)

```python
def find_non_unique_values(d: dict) -> dict:
    # First invert the dictionary
    inverted = {}
    for key, value in d.items():
        if value not in inverted:
            inverted[value] = []
        inverted[value].append(key)
    # Keep only values that have multiple keys
    return {value: keys for value, keys in inverted.items() if len(keys) > 1}
```

```python
# Test find_non_unique_values
assert find_non_unique_values({'a': 1, 'b': 1, 'c': 2}) == {1: ['a', 'b']}
assert find_non_unique_values({'a': 1, 'b': 2, 'c': 3}) == {}
assert find_non_unique_values({'x': 'y', 'z': 'y'}) == {'y': ['x', 'z']}
```
