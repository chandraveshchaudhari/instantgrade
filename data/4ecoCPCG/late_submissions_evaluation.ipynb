# Control Flow Practice Exam - Late Submissions Evaluation

This notebook evaluates all late student submissions from the "Late in class assignment" folder.

```python
import sys
from pathlib import Path

# current notebook is inside ./data/4ecoCPCG/
NOTEBOOK_DIR = Path().resolve()
PROJECT_ROOT = NOTEBOOK_DIR.parent.parent     # go up two levels to project root

sys.path.append(str(PROJECT_ROOT / "src"))

from instantgrade import Evaluator
import os

print(f"Notebook directory: {NOTEBOOK_DIR}")
print(f"Project root: {PROJECT_ROOT}")
```

## Setup: Ensure logs and reports directories exist

```python
# Create logs and reports directories if they don't exist
logs_dir = NOTEBOOK_DIR / "logs_late"
reports_dir = NOTEBOOK_DIR / "reports_late"

logs_dir.mkdir(exist_ok=True)
reports_dir.mkdir(exist_ok=True)

print(f"Logs directory: {logs_dir}")
print(f"Reports directory: {reports_dir}")
```

## Verify solution and submission files

```python
# Use the instructor solution file WITH tests
solution_file = NOTEBOOK_DIR / "sample_solutions.ipynb"
submission_folder = NOTEBOOK_DIR / "Late in class assignment"

# Check if files exist
print(f"Solution file exists: {solution_file.exists()}")
print(f"Submission folder exists: {submission_folder.exists()}")

if submission_folder.exists():
    notebook_files = [f for f in submission_folder.glob("*.ipynb") if f.is_file()]
    print(f"Found {len(notebook_files)} student notebook files")
    for nb in notebook_files[:5]:  # Show first 5
        print(f"  - {nb.name}")
    if len(notebook_files) > 5:
        print(f"  ... and {len(notebook_files) - 5} more")
```

## Run the Evaluator using local execution (no Docker)

```python
# Test with LOCAL execution first to see detailed output
evaluator = Evaluator(
    solution_file_path=str(solution_file),
    submission_folder_path=str(submission_folder),
    use_docker=False,      # Use local execution for debugging
    log_path=str(logs_dir),
    log_level="DEBUG",
)

print("Running evaluation for late submissions...")
report = evaluator.run()
print("Evaluation complete!")
```

## Generate HTML Report

```python
# Save HTML report
report_path = reports_dir / "evaluation_report_late.html"
evaluator.to_html(str(report_path))
print(f"Report saved to: {report_path}")
```

## Summary Statistics - Late Submissions

```python
# Display summary
print("\n" + "="*80)
print("LATE SUBMISSIONS EVALUATION SUMMARY")
print("="*80)

if hasattr(report, 'executed_results') and report.executed_results:
    total_students = len(report.executed_results)
    print(f"Total late submissions evaluated: {total_students}")

    # Calculate statistics
    total_passed = 0
    total_failed = 0

    for student_result in report.executed_results:
        if 'results' in student_result:
            for result in student_result['results']:
                if result.get('status') == 'passed':
                    total_passed += 1
                else:
                    total_failed += 1

    print(f"Total test assertions: {total_passed + total_failed}")
    print(f"Passed: {total_passed}")
    print(f"Failed: {total_failed}")
    if total_passed + total_failed > 0:
        pass_rate = (total_passed / (total_passed + total_failed)) * 100
        print(f"Pass rate: {pass_rate:.1f}%")

    print("\nDetailed Results (all late submissions):")
    print("-" * 80)

    for i, student_result in enumerate(report.executed_results, 1):
        student_path = student_result.get('student_path', 'Unknown')
        student_name = str(student_path).split('/')[-1].replace('.ipynb', '')

        if 'results' in student_result:
            passed = sum(1 for r in student_result['results'] if r.get('status') == 'passed')
            total = len(student_result['results'])
            score = sum(r.get('score', 0) for r in student_result['results'])
            max_score = total
            percentage = (score / max_score * 100) if max_score > 0 else 0
            print(f"{i:2d}. {student_name:40s} | {passed}/{total} | Score: {score}/{max_score} ({percentage:5.1f}%)")
        else:
            print(f"{i:2d}. {student_name:40s} | Execution error or skipped")
else:
    print("No results found in report.")
```

## Detailed inspection of first 3 late submissions

```python
# Display the full report (first 3 students)
if hasattr(report, 'executed_results') and report.executed_results:
    for student_result in report.executed_results[:3]:
        student_path = str(student_result.get('student_path', 'Unknown')).split('/')[-1]
        print(f"\n{'='*80}")
        print(f"Student: {student_path}")
        print('='*80)

        if 'results' in student_result:
            for result in student_result['results'][:3]:
                print(f"\n  Question: {result.get('question')}")
                print(f"  Status: {result.get('status')}")
                print(f"  Score: {result.get('score')}/{1}")
                print(f"  Assertion: {result.get('assertion')[:80]}...")
                if result.get('error'):
                    print(f"  Error: {str(result.get('error'))[:200]}...")
        else:
            print("  No results (execution error)")
```

## Comparison: On-Time vs Late Submissions

```python
print("\n" + "="*80)
print("COMPARISON: ON-TIME vs LATE SUBMISSIONS")
print("="*80)

on_time_stats = {
    'total_students': 26,
    'passed': 519,
    'failed': 27,
    'pass_rate': 95.1,
}

if hasattr(report, 'executed_results') and report.executed_results:
    late_students = len(report.executed_results)
    late_passed = 0
    late_failed = 0

    for student_result in report.executed_results:
        if 'results' in student_result:
            for result in student_result['results']:
                if result.get('status') == 'passed':
                    late_passed += 1
                else:
                    late_failed += 1

    late_pass_rate = (late_passed / (late_passed + late_failed) * 100) if (late_passed + late_failed) > 0 else 0

    late_stats = {
        'total_students': late_students,
        'passed': late_passed,
        'failed': late_failed,
        'pass_rate': late_pass_rate,
    }

    print(f"\n{'Metric':<25} {'On-Time':<20} {'Late':<20}")
    print("-" * 80)
    print(f"{'Total Students':<25} {on_time_stats['total_students']:<20} {late_stats['total_students']:<20}")
    print(f"{'Total Assertions':<25} {on_time_stats['passed'] + on_time_stats['failed']:<20} {late_stats['passed'] + late_stats['failed']:<20}")
    print(f"{'Passed':<25} {on_time_stats['passed']:<20} {late_stats['passed']:<20}")
    print(f"{'Failed':<25} {on_time_stats['failed']:<20} {late_stats['failed']:<20}")
    print(f"{'Pass Rate':<25} {on_time_stats['pass_rate']:.1f}%{'':<14} {late_stats['pass_rate']:.1f}%")

    print("\n" + "="*80)
    print("OBSERVATIONS")
    print("="*80)

    if late_pass_rate >= on_time_stats['pass_rate']:
        diff = late_pass_rate - on_time_stats['pass_rate']
        print(f"✅ Late submissions performed BETTER than on-time submissions (+{diff:.1f}%)")
    else:
        diff = on_time_stats['pass_rate'] - late_pass_rate
        print(f"⚠️  Late submissions performed WORSE than on-time submissions (-{diff:.1f}%)")

    print(f"\nLate submission rate: {late_students} students submitted late")
```
