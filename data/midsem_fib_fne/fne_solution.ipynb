### 1. Create a set from a list
**Return:** A set containing unique elements from the list

```python
def list_to_set(lst: list) -> set:
    return set(lst)
```

```python
assert list_to_set([1, 2, 3, 3, 2]) == {1, 2, 3}
assert list_to_set([]) == set()
assert list_to_set([5]) == {5}
```

### 2. Remove a specific element from a list
**Return:** The list after removing the first occurrence of the element

```python
def remove_element(lst: list, element) -> list:
    if element in lst:
        lst.remove(element)
    return lst
```

```python
assert remove_element([1, 2, 3, 2], 2) == [1, 3, 2]
assert remove_element([1, 2, 3], 5) == [1, 2, 3]
assert remove_element([5], 5) == []
```

### 3. Remove a key from a dictionary
**Return:** The dictionary after removing the key (if it exists)

```python
def remove_key(d: dict, key) -> dict:
    if key in d:
        del d[key]
    return d
```

```python
assert remove_key({'a': 1, 'b': 2, 'c': 3}, 'a') == {'b': 2, 'c': 3}
assert remove_key({'x': 10}, 'x') == {}
assert remove_key({'a': 1}, 'z') == {'a': 1}
```

### 4. Find factorial of a number using a for loop
**Return:** The factorial of n (n!)

```python
def factorial(n: int) -> int:
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
assert factorial(0) == 1
assert factorial(1) == 1
assert factorial(5) == 120
assert factorial(6) == 720
```

### 5. Find the intersection of two sets
**Return:** A set containing elements common to both sets

```python
def set_intersection(s1: set, s2: set) -> set:
    return s1.intersection(s2)
```

```python
assert set_intersection({1, 2, 3}, {2, 3, 4}) == {2, 3}
assert set_intersection({1, 2, 3}, {4, 5, 6}) == set()
assert set_intersection({1, 2, 3}, {1, 2, 3}) == {1, 2, 3}
```

### 6. Count vowels in a string using a for loop
**Return:** The number of vowels in the string

```python
def count_vowels(s: str) -> int:
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
```

```python
assert count_vowels("hello") == 2
assert count_vowels("aeiou") == 5
assert count_vowels("bcdfg") == 0
assert count_vowels("Happy World") == 3
```

### 7. Categorize temperature (Hot, Warm, Cold)
**Return:** 'Hot' if temp > 30, 'Warm' if 15-30, 'Cold' if < 15

```python
def categorize_temperature(temp: float) -> str:
    if temp > 30:
        return 'Hot'
    elif temp >= 15:
        return 'Warm'
    else:
        return 'Cold'
```

```python
assert categorize_temperature(35) == 'Hot'
assert categorize_temperature(30) == 'Warm'
assert categorize_temperature(20) == 'Warm'
assert categorize_temperature(15) == 'Warm'
assert categorize_temperature(10) == 'Cold'
assert categorize_temperature(31) == 'Hot'
```

### 8. Sort a list in ascending order
**Return:** The sorted list

```python
def sort_list(lst: list) -> list:
    return sorted(lst)
```

```python
assert sort_list([3, 1, 5, 2]) == [1, 2, 3, 5]
assert sort_list([]) == []
assert sort_list([1]) == [1]
assert sort_list([5, 4, 3, 2, 1]) == [1, 2, 3, 4, 5]
```

### 9. Create a list of squares from 1 to n
**Return:** A list of squares of numbers from 1 to n

```python
def list_of_squares(n: int) -> list:
    squares = []
    for i in range(1, n + 1):
        squares.append(i ** 2)
    return squares
```

```python
assert list_of_squares(5) == [1, 4, 9, 16, 25]
assert list_of_squares(1) == [1]
assert list_of_squares(0) == []
```

### 10. Slice a list from index start to end
**Return:** The sliced portion of the list

```python
def slice_list(lst: list, start: int, end: int) -> list:
    return lst[start:end]
```

```python
assert slice_list([1, 2, 3, 4, 5], 0, 3) == [1, 2, 3]
assert slice_list([1, 2, 3, 4, 5], 2, 5) == [3, 4, 5]
assert slice_list([1, 2, 3], 0, 0) == []
```
