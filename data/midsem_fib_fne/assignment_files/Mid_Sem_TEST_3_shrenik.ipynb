```python
def append_element(lst: list, element) -> list:
    if element not in lst:
        lst.append(element)
        return lst
    else:
        return lst
```

```python
def add_to_set(s: set, element) -> set:
    if element not in s:
        s.add(element)
        return s
    else :
        return s
```

```python
def remove_key(d: dict, key) -> dict:
    if key in d:
        del d[key]
        return d
    else:
        return d
    
```

```python
def multiplication_table(n: int) -> list:
    table =[]
    for i in range(1,11):
        table.append(n*i)
    return table
```

```python
def set_difference(s1: set, s2: set) -> set:
    if s1.issubclass(s2):
        return s1.difference(s2)
    else :
        return s2.difference(s1)
```

```python
def reverse_string(s: str) -> str:
    if s == s[::-1]:
        return True
    else:
        return False
```

```python
def sum_even_numbers(n: int) -> int:
    if n%2 == 0:
        sum = 0
        for i in range(1,n+1):
            if i%2 == 0:
                sum = sum + i
                return sum 
```
