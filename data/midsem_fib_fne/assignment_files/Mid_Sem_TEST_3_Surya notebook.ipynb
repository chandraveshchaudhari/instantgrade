### 1. Append an element to a list
**Return:** The list after appending the element

```python
def append_element(lst: list, F) -> list:
    return lst + [F]
    
```

### 2. Add an element to a set
**Return:** The set after adding the element

```python
def add_to_set(s: set, element) -> set:
    s.add(element)
    return s
```

### 3. Remove a key from a dictionary
**Return:** The dictionary after removing the key (if it exists)

```python
def remove_key(d: dict, key) -> dict:
    if key in d:
        del d[key]
    return d
```

### 4. Find the largest of three numbers
**Return:** The largest number among the three

```python
def largest_of_three(a: int, b: int, c: int) -> int:
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
```

### 5. Generate multiplication table for a number
**Return:** A list containing the multiplication table (1 to 10)

```python
def multiplication_table(n: int) -> list:
    return [n * i for i in range(1, 11)]
```

### 6. Find the difference between two sets
**Return:** A set containing elements in the first set but not in the second

```python
def set_difference(s1: set, s2: set) -> set:
    return s1.difference(s2)
```

### 7. Reverse a string using a for loop
**Return:** The reversed string

```python
def reverse_string(s: str) -> str:
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str
```

### 9. Create a dictionary from two lists (keys and values)
**Return:** A dictionary with keys from the first list and values from the second

```python
def lists_to_dict(keys: list, values: list) -> dict:
    return dict(zip(keys, values))
```

### 10. Determine ticket price based on age
**Return:** 0 for age < 3, 10 for 3-12, 15 for 13-64, 12 for 65+

```python
def ticket_price(age: int) -> int:
    if age < 3:
        return 0
    elif 3 <= age <= 12:
        return 10
    elif 13 <= age <= 64:
        return 15
    else: 
        return 12
```
