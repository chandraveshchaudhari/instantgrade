### 1. Create a set from a list
**Return:** A set containing unique elements from the list

```python
def list_to_set(lst: list) -> set:
    return set(lst)

# list_to_set([1, 2, 2, 3, 4, 4, 5])
```

```python
# Assertions
assert list_to_set([1, 2, 2, 3, 4, 4, 5]) == {1, 2, 3, 4, 5}
assert list_to_set(['a', 'b', 'a', 'c']) == {'a', 'b', 'c'}
assert list_to_set([]) == set()
```

### 2. Add an element to a set
**Return:** The set after adding the element

```python
def add_to_set(s: set, element) -> set:
    s.add(element)
    return s

# add_to_set({1, 2, 3}, 4)
```

```python
assert add_to_set({1, 2, 3}, 4) == {1, 2, 3, 4}
assert add_to_set({'a', 'b'}, 'c') == {'a', 'b', 'c'}
assert add_to_set({1, 2}, 2) == {1, 2}
```

### 3. Remove an element from a set (if it exists)
**Return:** The set after removing the element

```python
def remove_from_set(s: set, element) -> set:
    s.discard(element)
    return s

# remove_from_set({1, 2, 3, 4}, 3)
```

```python
assert remove_from_set({1, 2, 3, 4}, 3) == {1, 2, 4}
assert remove_from_set({'a', 'b', 'c'}, 'b') == {'a', 'c'}
assert remove_from_set({1, 2, 3}, 5) == {1, 2, 3}
```

### 4. Find the union of two sets
**Return:** A set containing all unique elements from both sets

```python
def set_union(s1: set, s2: set) -> set:
    return s1.union(s2)

# set_union({1, 2, 3}, {3, 4, 5})
```

```python
assert set_union({1, 2, 3}, {3, 4, 5}) == {1, 2, 3, 4, 5}
assert set_union({'a', 'b'}, {'b', 'c'}) == {'a', 'b', 'c'}
assert set_union({1}, {2}) == {1, 2}
```

### 5. Find the intersection of two sets
**Return:** A set containing elements common to both sets

```python
def set_intersection(s1: set, s2: set) -> set:
    return s1.intersection(s2)

# set_intersection({1, 2, 3}, {2, 3, 4})
```

```python
assert set_intersection({1, 2, 3}, {2, 3, 4}) == {2, 3}
assert set_intersection({'a', 'b', 'c'}, {'b', 'c', 'd'}) == {'b', 'c'}
assert set_intersection({1, 2}, {3, 4}) == set()
```

### 6. Find the difference between two sets
**Return:** A set containing elements in the first set but not in the second

```python
def set_difference(s1: set, s2: set) -> set:
    return s1.difference(s2)

# set_difference({1, 2, 3, 4}, {3, 4, 5})
```

```python
assert set_difference({1, 2, 3, 4}, {3, 4, 5}) == {1, 2}
assert set_difference({'a', 'b', 'c'}, {'b', 'd'}) == {'a', 'c'}
assert set_difference({1, 2}, {3, 4}) == {1, 2}
```

### 7. Check if an element exists in a set
**Return:** True if the element is in the set, False otherwise

```python
def element_in_set(s: set, element) -> bool:
    return element in s

# element_in_set({1, 2, 3, 4}, 3)
```

```python
assert element_in_set({1, 2, 3, 4}, 3) == True
assert element_in_set({'a', 'b', 'c'}, 'd') == False
assert element_in_set({10, 20}, 10) == True
```

### 8. Get the size of a set
**Return:** The number of elements in the set

```python
def set_size(s: set) -> int:
    return len(s)

# set_size({1, 2, 3, 4, 5})
```

```python
assert set_size({1, 2, 3, 4, 5}) == 5
assert set_size(set()) == 0
assert set_size({'a'}) == 1
```

### 9. Check if one set is a subset of another
**Return:** True if the first set is a subset of the second, False otherwise

```python
def is_subset(s1: set, s2: set) -> bool:
    return s1.issubset(s2)

# is_subset({1, 2}, {1, 2, 3, 4})
```

```python
assert is_subset({1, 2}, {1, 2, 3, 4}) == True
assert is_subset({'a', 'b'}, {'a', 'b', 'c'}) == True
assert is_subset({1, 5}, {1, 2, 3}) == False
```

### 10. Clear all elements from a set
**Return:** An empty set

```python
def clear_set(s: set) -> set:
    s.clear()
    return s

# clear_set({1, 2, 3, 4})
```

```python
assert clear_set({1, 2, 3, 4}) == set()
assert clear_set({'a', 'b', 'c'}) == set()
assert clear_set(set()) == set()
```
