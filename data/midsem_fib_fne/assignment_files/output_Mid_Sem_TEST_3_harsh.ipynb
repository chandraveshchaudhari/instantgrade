### 5. Generate multiplication table for a number
**Return:** A list containing the multiplication table (1 to 10)

```python
def multiplication_table(n: int) -> list:
    pass
```

```python
def generate_multiplication_table(number):
    """
    Generates a multiplication table for a given number (1 to 10).

    Args:
        number: The integer for which to generate the table.

    Returns:
        A list containing the products (e.g., [number*1, number*2, ...]).
    """
    table_list = []
    # Loop from 1 up to and including 10 (range stops before the second argument)
    for i in range(1, 11):
        result = number * i
        table_list.append(result)
    return table_list

# Example usage:
num = 5
table = generate_multiplication_table(num)
print(f"Multiplication table of {num}: {table}")
```

### 7. Reverse a string using a for loop
**Return:** The reversed string

```python
def reverse_string(s: str) -> str:
    pass
```

```python
def reverse_string_for_loop(s):
    reversed_s = ""
    for char in s:
        reversed_s = char + reversed_s
    return reversed_s

# Example Usage:
original_string = "Python"
reversed_string = reverse_string_for_loop(original_string)
print(f"Original String: {original_string}")
print(f"Reversed String: {reversed_string}")
```

### 8. Find sum of even numbers from 1 to n
**Return:** The sum of all even numbers from 1 to n

```python
def sum_even_numbers(n: int) -> int:
    pass
```

```python
def sum_even_numbers(n: int) -> int:
    total = 0
    for i in range(1, n + 1):
        if i % 2 == 0:
            total += i
    return total
```

### 10. Determine ticket price based on age
**Return:** 0 for age < 3, 10 for 3-12, 15 for 13-64, 12 for 65+

```python
def ticket_price(age: int) -> int:
    pass
```

```python
def get_ticket_price(age):
    if age < 3:
        return 0
    elif 3 <= age <= 12:
        return 10
    elif 13 <= age <= 64:
        return 15
    elif age >= 65:
        return 12
    else:
        return "Invalid Age"
```
