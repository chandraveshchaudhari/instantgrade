### 1. Get the value associated with a key
**Return:** The value for the given key, or None if the key doesn't exist

```python
def get_value(d: dict, key) -> any:
    return d.get(key)

# get_value({'name': 'Alice', 'age': 25}, 'name')
```

```python
# Assertions
assert get_value({'name': 'Alice', 'age': 25}, 'name') == 'Alice'
assert get_value({'a': 1, 'b': 2}, 'b') == 2
assert get_value({'x': 10}, 'y') == None
```

### 2. Add a new key-value pair to a dictionary
**Return:** The dictionary after adding the new pair

```python
def add_key_value(d: dict, key, value) -> dict:
    d[key] = value
    return d

# add_key_value({'a': 1}, 'b', 2)
```

```python
assert add_key_value({'a': 1}, 'b', 2) == {'a': 1, 'b': 2}
assert add_key_value({}, 'name', 'Bob') == {'name': 'Bob'}
assert add_key_value({'x': 10}, 'x', 20) == {'x': 20}
```

### 3. Remove a key from a dictionary
**Return:** The dictionary after removing the key (if it exists)

```python
def remove_key(d: dict, key) -> dict:
    if key in d:
        del d[key]
    return d

# remove_key({'a': 1, 'b': 2, 'c': 3}, 'b')
```

```python
assert remove_key({'a': 1, 'b': 2, 'c': 3}, 'b') == {'a': 1, 'c': 3}
assert remove_key({'name': 'Alice', 'age': 25}, 'age') == {'name': 'Alice'}
assert remove_key({'x': 10}, 'y') == {'x': 10}
```

### 4. Check if a key exists in a dictionary
**Return:** True if the key exists, False otherwise

```python
def key_exists(d: dict, key) -> bool:
    return key in d

# key_exists({'a': 1, 'b': 2}, 'a')
```

```python
assert key_exists({'a': 1, 'b': 2}, 'a') == True
assert key_exists({'name': 'Alice'}, 'age') == False
assert key_exists({}, 'x') == False
```

### 5. Get all keys from a dictionary
**Return:** A list of all keys in the dictionary

```python
def get_keys(d: dict) -> list:
    return list(d.keys())

# get_keys({'a': 1, 'b': 2, 'c': 3})
```

```python
assert sorted(get_keys({'a': 1, 'b': 2, 'c': 3})) == ['a', 'b', 'c']
assert get_keys({}) == []
assert sorted(get_keys({'x': 10, 'y': 20})) == ['x', 'y']
```

### 6. Get all values from a dictionary
**Return:** A list of all values in the dictionary

```python
def get_values(d: dict) -> list:
    return list(d.values())

# get_values({'a': 1, 'b': 2, 'c': 3})
```

```python
assert sorted(get_values({'a': 1, 'b': 2, 'c': 3})) == [1, 2, 3]
assert get_values({}) == []
assert sorted(get_values({'x': 10, 'y': 20})) == [10, 20]
```

### 7. Update a dictionary with another dictionary
**Return:** The updated dictionary

```python
def update_dict(d1: dict, d2: dict) -> dict:
    d1.update(d2)
    return d1

# update_dict({'a': 1, 'b': 2}, {'b': 3, 'c': 4})
```

```python
assert update_dict({'a': 1, 'b': 2}, {'b': 3, 'c': 4}) == {'a': 1, 'b': 3, 'c': 4}
assert update_dict({}, {'x': 10}) == {'x': 10}
assert update_dict({'name': 'Alice'}, {'age': 25}) == {'name': 'Alice', 'age': 25}
```

### 8. Get the number of key-value pairs in a dictionary
**Return:** The number of items in the dictionary

```python
def dict_length(d: dict) -> int:
    return len(d)

# dict_length({'a': 1, 'b': 2, 'c': 3})
```

```python
assert dict_length({'a': 1, 'b': 2, 'c': 3}) == 3
assert dict_length({}) == 0
assert dict_length({'x': 10}) == 1
```

### 9. Create a dictionary from two lists (keys and values)
**Return:** A dictionary with keys from the first list and values from the second

```python
def lists_to_dict(keys: list, values: list) -> dict:
    return dict(zip(keys, values))

# lists_to_dict(['a', 'b', 'c'], [1, 2, 3])
```

```python
assert lists_to_dict(['a', 'b', 'c'], [1, 2, 3]) == {'a': 1, 'b': 2, 'c': 3}
assert lists_to_dict(['name', 'age'], ['Alice', 25]) == {'name': 'Alice', 'age': 25}
assert lists_to_dict([], []) == {}
```

### 10. Get all items from a dictionary as a list of tuples
**Return:** A list of (key, value) tuples

```python
def get_items(d: dict) -> list:
    return list(d.items())

# get_items({'a': 1, 'b': 2})
```

```python
assert sorted(get_items({'a': 1, 'b': 2})) == [('a', 1), ('b', 2)]
assert get_items({}) == []
assert sorted(get_items({'x': 10, 'y': 20})) == [('x', 10), ('y', 20)]
```
