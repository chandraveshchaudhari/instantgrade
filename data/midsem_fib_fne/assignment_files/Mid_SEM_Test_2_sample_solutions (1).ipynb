### 1. Append an element to a list
**Return:** The list after appending the element

```python
def append_element(lst: list, element) -> list:
    lst.append(element)
    return lst

# append_element([1, 2, 3], 4)
```

```python
# Assertions
assert append_element([1, 2, 3], 4) == [1, 2, 3, 4]
assert append_element(['a', 'b'], 'c') == ['a', 'b', 'c']
assert append_element([], 1) == [1]
```

### 2. Remove a specific element from a list
**Return:** The list after removing the first occurrence of the element

```python
def remove_element(lst: list, element) -> list:
    if element in lst:
        lst.remove(element)
    return lst

# remove_element([1, 2, 3, 2], 2)
```

```python
assert remove_element([1, 2, 3, 2], 2) == [1, 3, 2]
assert remove_element(['a', 'b', 'c'], 'b') == ['a', 'c']
assert remove_element([1, 2, 3], 5) == [1, 2, 3]
```

### 3. Insert an element at a specific index
**Return:** The list after inserting the element at the given index

```python
def insert_element(lst: list, index: int, element) -> list:
    lst.insert(index, element)
    return lst

# insert_element([1, 2, 4], 2, 3)
```

```python
assert insert_element([1, 2, 4], 2, 3) == [1, 2, 3, 4]
assert insert_element(['a', 'c'], 1, 'b') == ['a', 'b', 'c']
assert insert_element([10, 20], 0, 5) == [5, 10, 20]
```

### 4. Get the length of a list
**Return:** The number of elements in the list

```python
def list_length(lst: list) -> int:
    return len(lst)

# list_length([1, 2, 3, 4, 5])
```

```python
assert list_length([1, 2, 3, 4, 5]) == 5
assert list_length([]) == 0
assert list_length(['a']) == 1
```

### 5. Find the maximum element in a list
**Return:** The largest element in the list

```python
def max_element(lst: list) -> int:
    return max(lst)

# max_element([3, 7, 2, 9, 1])
```

```python
assert max_element([3, 7, 2, 9, 1]) == 9
assert max_element([10]) == 10
assert max_element([-5, -2, -10]) == -2
```

### 6. Find the minimum element in a list
**Return:** The smallest element in the list

```python
def min_element(lst: list) -> int:
    return min(lst)

# min_element([3, 7, 2, 9, 1])
```

```python
assert min_element([3, 7, 2, 9, 1]) == 1
assert min_element([10]) == 10
assert min_element([-5, -2, -10]) == -10
```

### 7. Reverse a list
**Return:** The list in reverse order

```python
def reverse_list(lst: list) -> list:
    return lst[::-1]

# reverse_list([1, 2, 3, 4, 5])
```

```python
assert reverse_list([1, 2, 3, 4, 5]) == [5, 4, 3, 2, 1]
assert reverse_list(['a', 'b', 'c']) == ['c', 'b', 'a']
assert reverse_list([10]) == [10]
```

### 8. Sort a list in ascending order
**Return:** The sorted list

```python
def sort_list(lst: list) -> list:
    return sorted(lst)

# sort_list([5, 2, 8, 1, 9])
```

```python
assert sort_list([5, 2, 8, 1, 9]) == [1, 2, 5, 8, 9]
assert sort_list([3, 1, 2]) == [1, 2, 3]
assert sort_list([10]) == [10]
```

### 9. Count occurrences of an element in a list
**Return:** The number of times the element appears in the list

```python
def count_occurrences(lst: list, element) -> int:
    return lst.count(element)

# count_occurrences([1, 2, 2, 3, 2, 4], 2)
```

```python
assert count_occurrences([1, 2, 2, 3, 2, 4], 2) == 3
assert count_occurrences(['a', 'b', 'a', 'c'], 'a') == 2
assert count_occurrences([1, 2, 3], 5) == 0
```

### 10. Slice a list from index start to end
**Return:** The sliced portion of the list

```python
def slice_list(lst: list, start: int, end: int) -> list:
    return lst[start:end]

# slice_list([1, 2, 3, 4, 5], 1, 4)
```

```python
assert slice_list([1, 2, 3, 4, 5], 1, 4) == [2, 3, 4]
assert slice_list(['a', 'b', 'c', 'd'], 0, 2) == ['a', 'b']
assert slice_list([10, 20, 30], 1, 3) == [20, 30]
```
