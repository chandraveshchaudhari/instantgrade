```python
my_list = [1, 2, 3]
my_list.append(4)
print(my_list) # Output: [1, 2, 3, 4]
```

```python
my_set = {1, 2, 3}
my_set.add(4)
print(my_set) 
# Output: {1, 2, 3, 4}
```

```python
my_dict = {"a": 1, "b": 2}
my_dict.pop("b", None)
my_dict.pop("c", None) 
print(my_dict)  # Output: {'a': 1}
```

```python
def find_largest(a, b, c):
    return max(a, b, c)
```

```python
def generate_multiplication_table(n):
    """Returns a list of the multiplication table for n from 1 to 10."""
    return [n * i for i in range(1, 11)]
number = 7
print(generate_multiplication_table(number))
```

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
result = set_a - set_b
print(result)  # Output: {1, 2}
```

```python
def reverse_string(s):
    reversed_s = ""
    for char in s:
       
        reversed_s = char + reversed_s
    return reversed_s
original = "Python"
print(reverse_string(original)
```
