### 4. Find factorial of a number using a for loop
**Return:** The factorial of n (n!)

```python
def factorial(n: int) -> int:
    pass
```

```python
def factorial(n):
   
    result = 1
    
    
    for i in range(1, n + 1):
        result *= i 
    
    return result


number = 67
print(f"The factorial of {number} is {factorial(number)}")
```

### 6. Count vowels in a string using a for loop
**Return:** The number of vowels in the string

```python
def count_vowels(s: str) -> int:
    pass
```

```python
def count_vowels(string):
   
    vowel_count = 0
    
    
    vowels = "aeiouAEIOU"
    
   
    for char in string:
        
        if char in vowels:
            vowel_count += 1
    
    return vowel_count


text = "Hello, World!"
print(f"The string '{text}' has {count_vowels(text)} vowels.")
```

### 7. Categorize temperature (Hot, Warm, Cold)
**Return:** 'Hot' if temp > 30, 'Warm' if 15-30, 'Cold' if < 15

```python
def categorize_temperature(temp: float) -> str:
    pass
```

```python
def categorize_temperature(temp):
    """
    Categorize temperature as Hot, Warm, or Cold
    
    Args:
        temp: Temperature in degrees (assumed to be Celsius)
        
    Returns:
        String describing the temperature category
    """
    if temp >= 30:
        return "Hot"
    elif 15 <= temp < 30:
        return "Warm"
    else:
        return "Cold"


temperature = 25
category = categorize_temperature(temperature)
print(f"A temperature of {temperature}°C is categorized as: {category}")
```
