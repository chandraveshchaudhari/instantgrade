### 4. Find factorial of a number using a for loop
**Return:** The factorial of n (n!)

```python
def factorial(n):
    prod=1
    for i in range(1,n+1):
        prod=prod*i
    return prod
```

```python
factorial(15)
```

### 6. Count vowels in a string using a for loop
**Return:** The number of vowels in the string

```python
def count_vowels(s):
    vowels="aeiouAEIOU"
    count=0
    for char in s:
        if char in vowels:
            count=count+1
    return count
```

```python
count_vowels("helloooo")
```

### 7. Categorize temperature (Hot, Warm, Cold)
**Return:** 'Hot' if temp > 30, 'Warm' if 15-30, 'Cold' if < 15

```python
def categorize_temperature(temp):
    if temp>30:
        return "hot"
    elif temp>=15 & temp<=30:
        return "warm"
    else:
        return "cold"
```

```python
categorize_temperature(35)
```
