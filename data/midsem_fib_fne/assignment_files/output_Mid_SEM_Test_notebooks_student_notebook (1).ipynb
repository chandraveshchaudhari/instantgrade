### 2. Remove a specific element from a list
**Return:** The list after removing the first occurrence of the element

```python
def remove_element(lst: list, element) -> list:
    pass
    if element in lst:
        lst.remove(element)
    return lst
numbers = [1, 2, 3, 2, 4]
print(remove_element(numbers, 2))

```

### 4. Find factorial of a number using a for loop
**Return:** The factorial of n (n!)

```python
def factorial(n: int) -> int:
    pass
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    
    result = 1
    for i in range(1, n + 1):
        result*= i
    
    return result
print(factorial(5))
```

### 5. Find the intersection of two sets
**Return:** A set containing elements common to both sets

```python
def set_intersection(s1: set, s2: set) -> set:
    pass
    result = set()
    for element in s1:
        if element in s2:
            result.add(element)
    return result
print(set_intersection({1, 2, 3}, {2, 3, 4}))
```

### 6. Count vowels in a string using a for loop
**Return:** The number of vowels in the string

```python
def count_vowels(s: str) -> int:
    pass
    vowels = "aeiouAEIOU"
    count = 0
    
    for char in s:
        if char in vowels:
            count += 1
    
    return count
print(count_vowels("my name is arjun "))
```

### 7. Categorize temperature (Hot, Warm, Cold)
**Return:** 'Hot' if temp > 30, 'Warm' if 15-30, 'Cold' if < 15

```python
def categorize_temperature(temp: float) -> str:
    pass
    if temp > 30:
        return "Hot"
    elif 15 <= temp <= 30:
        return "Warm"
    else:
        return "Cold"
print(categorize_temperature(10))
print(categorize_temperature(25))
print(categorize_temperature(35))
```

### 8. Sort a list in ascending order
**Return:** The sorted list

```python
def sort_list(lst: list) -> list:
    pass
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] > lst[j]:
                lst[i], lst[j] = lst[j], lst[i]
    return lst
print(sort_list([5, 2, 9, 1]))
```

### 9. Create a list of squares from 1 to n
**Return:** A list of squares of numbers from 1 to n

```python
def list_of_squares(n: int) -> list:
    pass
    squares = []
    
    for i in range(1, n + 1):
        squares.append(i * i)
    return squares
print(list_of_squares(5))
```
