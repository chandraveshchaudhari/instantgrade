### 10. Determine ticket price based on age

```python
def ticket_price(age: int) -> int:
    pass

def ticket_price(age: int) -> int:
    """Determine ticket price based on age.
    0 for age < 3, 10 for 3-12, 15 for 13-64, 12 for 65+
    """
    if age < 3:
        return 0
    elif 3 <= age <= 12:
        return 10
    elif 13 <= age <= 64:
        return 15
    else:  # age >= 65
        return 12

        
```
