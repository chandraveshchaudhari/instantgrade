### 9. Rename Column – James Bond Dataset

Return: A DataFrame with column `Bond Actor Salary` renamed to `Bond_Salary`.

Rename the column `Bond Actor Salary` to `Bond_Salary` in the `bond` DataFrame.

```python
def rename_bond_salary(df: pd.DataFrame) -> pd.DataFrame:
    return df.rename(columns={'Bond Actor Salary': 'Bond_Salary'})
```

### 10. Data Type Conversion – James Bond Dataset

Return: A DataFrame where the `Bond_Salary` column is of numeric dtype (non-convertible values coerced to NaN).

Convert the `Bond_Salary` column to numeric using `errors='coerce'`.

```python
def convert_bond_salary_to_numeric(df: pd.DataFrame) -> pd.DataFrame:
    df['Bond_Salary'] = pd.to_numeric(df['Bond_Salary'], errors='coerce')
    return df
```

```python
bond_renamed = rename_bond_salary(bond)
result = convert_bond_salary_to_numeric(bond_renamed)
assert pd.api.types.is_numeric_dtype(result['Bond_Salary']), "Bond_Salary should be numeric"
```

### 11. Handling Missing Values – James Bond Dataset

Return: A DataFrame where missing values in `Bond_Salary` are replaced by 0.

Replace all missing values in `Bond_Salary` with `0`.

```python
def fill_bond_salary_zero(df: pd.DataFrame) -> pd.DataFrame:
    df['Bond_Salary'] = df['Bond_Salary'].fillna(0)
    return df
```

```python
bond_renamed = rename_bond_salary(bond)
bond_numeric = convert_bond_salary_to_numeric(bond_renamed)
result = fill_bond_salary_zero(bond_numeric)
assert result['Bond_Salary'].isna().sum() == 0, "No missing values should remain after fillna"
```

### 12. Date Conversion – Google Stock Dataset

### 15. Filtering Data – Google Stock Dataset

Return: DataFrame containing rows where `Price` is greater than the average `Price`.

Display all rows from the `google` DataFrame where the `Price` is greater than the average `Price`.

```python
def filter_price_above_average(df: pd.DataFrame) -> pd.DataFrame:
    average_price = df['Price'].mean()
    return df[df['Price'] > average_price]
```

```python
result = filter_price_above_average(google)
assert isinstance(result, pd.DataFrame), "filter_price_above_average should return a DataFrame"
assert all(result['Price'] > google['Price'].mean()), "All prices should be above average"
```

### 16. Grouping Data – James Bond Dataset

Return: A Series mapping each actor to the number of films they performed in the dataset.

Display the number of films performed by each actor using the `bond` DataFrame.

```python
def films_per_actor(df: pd.DataFrame, actor_col: str='Actor') -> pd.Series:
    return df[actor_col].value_counts()
```

```python
result = films_per_actor(bond)
assert isinstance(result, pd.Series), "films_per_actor should return a Series"
assert len(result) > 0, "Series should not be empty"
```
