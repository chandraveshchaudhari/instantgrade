### 10. Data Type Conversion – James Bond Dataset

Return: A DataFrame where the `Bond_Salary` column is of numeric dtype (non-convertible values coerced to NaN).

Convert the `Bond_Salary` column to numeric using `errors='coerce'`.

```python
def convert_bond_salary_to_numeric(df: pd.DataFrame) -> pd.DataFrame:
    import pandas as pd


    df['Bond_Salary'] = pd.to_numeric(df['Bond_Salary'], errors='coerce')
    return df
```

### 11. Handling Missing Values – James Bond Dataset

Replace all missing values in `Bond_Salary` with `0`.

```python
def fill_bond_salary_zero(df: pd.DataFrame) -> pd.DataFrame:
    import pandas as pd


    df['Bond_Salary'] = df['Bond_Salary'].fillna(0)
    return df
```

### 15. Filtering Data – Google Stock Dataset

Return: DataFrame containing rows where `Price` is greater than the average `Price`.

Display all rows from the `google` DataFrame where the `Price` is greater than the average `Price`.

```python
def filter_price_above_average(df: pd.DataFrame) -> pd.DataFrame:
    import pandas as pd


    # Calculate the mean of the Price column
    average_price = df['Price'].mean()
    
    # Return a new DataFrame with boolean indexing applied
    return df[df['Price'] > average_price]
```
