### 16. Grouping Data – James Bond Dataset

Return: A Series mapping each actor to the number of films they performed in the dataset.

Display the number of films performed by each actor using the `bond` DataFrame.

```python
def films_per_actor(df: pd.DataFrame, actor_col: str='Actor') -> pd.Series:
    result=df.groupby('Actor').agg(total_films=('Films','sum'))
    return result
```
