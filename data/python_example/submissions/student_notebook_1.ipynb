```python
def load_bond(path_or_buffer: str='jamesbond - jamesbond.csv') -> pd.DataFrame:
    return pd.read_csv(path_or_buffer)

bond = load_bond()
assert isinstance(bond, pd.DataFrame)
assert len(bond) > 0
```

```python
def load_google(path_or_buffer: str='google_stock_price - google_stock_price.csv') -> pd.DataFrame:
    return pd.read_csv(path_or_buffer)

google = load_google()
assert isinstance(google, pd.DataFrame)
assert len(google) > 0
```

```python
def bond_head(df: pd.DataFrame) -> pd.DataFrame:
    return df.head()

result = bond_head(bond)
assert isinstance(result, pd.DataFrame)
assert len(result) <= 5
```

```python
def google_head(df: pd.DataFrame) -> pd.DataFrame:
    return df.head()

result = google_head(google)
assert isinstance(result, pd.DataFrame)
assert len(result) <= 5
```

```python
def bond_columns(df: pd.DataFrame) -> list:
    return df.columns.tolist()

result = bond_columns(bond)
assert isinstance(result, list)
assert len(result) > 0
```

```python
def google_columns(df: pd.DataFrame) -> list:
    return df.columns.tolist()

result = google_columns(google)
assert isinstance(result, list)
assert len(result) > 0
```

```python
def bond_shape(df: pd.DataFrame) -> tuple:
    return df.shape

result = bond_shape(bond)
assert isinstance(result, tuple)
assert len(result) == 2
```

```python
def google_shape(df: pd.DataFrame) -> tuple:
    return df.shape

result = google_shape(google)
assert isinstance(result, tuple)
assert len(result) == 2
```

```python
def rename_bond_salary(df: pd.DataFrame) -> pd.DataFrame:
    return df.rename(columns={'Bond Actor Salary': 'Bond_Salary'})

result = rename_bond_salary(bond)
assert 'Bond_Salary' in result.columns
assert 'Bond Actor Salary' not in result.columns
```

```python
def convert_bond_salary_to_numeric(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df['Bond_Salary'] = pd.to_numeric(df['Bond_Salary'], errors='coerce')
    return df

bond_renamed = rename_bond_salary(bond)
result = convert_bond_salary_to_numeric(bond_renamed)
import pandas as _pd
assert _pd.api.types.is_numeric_dtype(result['Bond_Salary'])
```

```python
def fill_bond_salary_zero(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df['Bond_Salary'] = df['Bond_Salary'].fillna(0)
    return df

bond_renamed = rename_bond_salary(bond)
bond_numeric = convert_bond_salary_to_numeric(bond_renamed)
result = fill_bond_salary_zero(bond_numeric)
assert result['Bond_Salary'].isna().sum() == 0
```

```python
def convert_google_date(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df['Date'] = pd.to_datetime(df['Date'])
    return df

result = convert_google_date(google.copy())
import pandas as _pd
assert _pd.api.types.is_datetime64_any_dtype(result['Date'])
```

```python
def set_google_index_date(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df = df.set_index('Date')
    return df

google_dated = convert_google_date(google.copy())
result = set_google_index_date(google_dated)
assert result.index.name == 'Date'
assert 'Date' not in result.columns
```
