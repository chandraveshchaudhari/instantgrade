## 1. Logistic Regression

Logistic Regression is a **discriminative probabilistic model** used for binary classification.

### Key Ideas
- Uses the **sigmoid function** to map outputs to probabilities
- Decision boundary is linear
- Outputs probability of class = 1

### Sigmoid Function
$$ \sigma(z) = \frac{1}{1 + e^{-z}} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([0,0,0,1,1])

model = LogisticRegression()
model.fit(X, y)

x_test = np.linspace(0,6,100).reshape(-1,1)
y_prob = model.predict_proba(x_test)[:,1]

plt.plot(x_test, y_prob)
plt.xlabel('Feature')
plt.ylabel('Probability of Class 1')
plt.title('Logistic Regression Probability Curve')
plt.show()
```

## 2. Naive Bayes

Naive Bayes is a **probabilistic generative model** based on Bayes' Theorem.

### Assumptions
- Features are conditionally independent
- Fast and efficient for large datasets

### Bayes Theorem
$$ P(y|X) = \frac{P(X|y)P(y)}{P(X)} $$

Common variants:
- Gaussian Naive Bayes
- Multinomial Naive Bayes (text data)

```python
from sklearn.naive_bayes import GaussianNB

X = np.array([[1,20],[2,21],[3,22],[4,10],[5,9]])
y = np.array([1,1,1,0,0])

model = GaussianNB()
model.fit(X, y)

print('Predictions:', model.predict(X))
```

## 3. Calibration & Class Imbalance

### Probability Calibration
- Ensures predicted probabilities reflect true likelihood
- Important for decision-making systems

### Class Imbalance
- One class dominates the dataset
- Leads to misleading accuracy

### Solutions
- Class weighting
- Resampling (oversampling / undersampling)
- Use metrics like Precision, Recall, ROC-AUC

```python
from sklearn.utils.class_weight import compute_class_weight

y = np.array([0,0,0,0,1])
weights = compute_class_weight('balanced', classes=np.unique(y), y=y)
print('Class Weights:', weights)
```

## 4. Lab – Churn Prediction

### Problem
Predict whether a customer will churn (1) or not (0).

### Steps
1. Train-test split
2. Train Logistic Regression
3. Evaluate performance

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

X = np.array([[25,500],[45,200],[30,300],[50,100],[23,700],[40,150]])
y = np.array([0,1,0,1,0,1])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Summary

- Logistic Regression is discriminative and probabilistic
- Naive Bayes is generative and fast
- Calibration improves probability reliability
- Class imbalance must be handled carefully
