# Performance Metrics & Visualisation

## Learning Objectives
- Understand regression performance metrics
- Understand classification performance metrics
- Learn how metrics are used in business decision-making
- Build a simple metric dashboard


## 1. Regression Metrics

Regression metrics measure how close predicted values are to actual values.

### Common Metrics
- **Mean Absolute Error (MAE)**: Average absolute difference
- **Mean Squared Error (MSE)**: Penalizes large errors
- **Root Mean Squared Error (RMSE)**: Square root of MSE
- **R² Score**: Proportion of variance explained by the model


```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

print('MAE:', mean_absolute_error(y_true, y_pred))
print('MSE:', mean_squared_error(y_true, y_pred))
print('RMSE:', mean_squared_error(y_true, y_pred, squared=False))
print('R2 Score:', r2_score(y_true, y_pred))
```

## 2. Classification Metrics

Classification metrics evaluate how well a model predicts class labels.

### Key Metrics
- **Accuracy**: Overall correctness
- **Precision**: Correct positive predictions
- **Recall**: Ability to find actual positives
- **F1 Score**: Balance between precision and recall
- **Confusion Matrix**: Detailed prediction breakdown


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

y_true = [1, 0, 1, 1, 0]
y_pred = [1, 0, 0, 1, 1]

print('Accuracy:', accuracy_score(y_true, y_pred))
print('Precision:', precision_score(y_true, y_pred))
print('Recall:', recall_score(y_true, y_pred))
print('F1 Score:', f1_score(y_true, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_true, y_pred))
```

## 3. Business Visualisation

Visualisation helps translate metrics into business insights.

### Why it matters
- Makes trends easy to understand
- Supports data-driven decisions
- Helps communicate results to non-technical stakeholders


```python
import matplotlib.pyplot as plt

metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
values = [0.8, 0.75, 0.7, 0.72]

plt.bar(metrics, values)
plt.ylim(0,1)
plt.title('Model Performance Metrics')
plt.ylabel('Score')
plt.show()
```

## 4. Lab – Metric Dashboard

This lab combines metrics and visualisation into a simple dashboard-like view.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,5,4,5])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('MAE:', mean_absolute_error(y_test, y_pred))
print('RMSE:', mean_squared_error(y_test, y_pred, squared=False))
print('R2:', r2_score(y_test, y_pred))
```

## Summary

- Classification metrics evaluate label accuracy
