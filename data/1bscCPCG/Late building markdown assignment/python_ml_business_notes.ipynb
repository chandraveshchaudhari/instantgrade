# Working with APIs

## What is an API?

An <mark style="background-color: #808000; color: white;">API (Application Programming Interface)</mark> is a set of rules and protocols that allows different software applications to communicate with each other. APIs enable developers to access data and functionality from external services programmatically.

## Types of APIs

### REST APIs

- **Most Common**: <mark style="background-color: #808000; color: white;">REpresentational State Transfer</mark>
- **HTTP Methods**: GET (retrieve), POST (create), PUT (update), DELETE (remove)
- **Stateless**: Each request contains all necessary information
- **Returns**: Usually JSON or XML format

### Other API Types

- **SOAP APIs**: XML-based protocol for enterprise applications
- **GraphQL**: Query language for APIs with flexible data retrieval
- **WebSocket APIs**: Real-time bidirectional communication

## Working with APIs in Python

### Essential Libraries

1. **requests**: <mark style="background-color: #808000; color: white;">Most popular library for making HTTP requests</mark>
2. **urllib**: Built-in Python library for URL handling
3. **httpx**: Modern async-capable HTTP client
4. **aiohttp**: Asynchronous HTTP client/server framework

### Making API Requests

#### GET Request Example

```python
import requests

# Basic GET request
response = requests.get('https://api.example.com/data')
data = response.json()

# GET request with parameters
params = {'key': 'value', 'limit': 10}
response = requests.get('https://api.example.com/data', params=params)
```

#### POST Request Example

```python
# POST request with JSON data
payload = {'name': 'John', 'age': 30}
response = requests.post('https://api.example.com/users', json=payload)
```

### Authentication

#### API Key Authentication

```python
headers = {'Authorization': 'Bearer YOUR_API_KEY'}
response = requests.get('https://api.example.com/data', headers=headers)
```

#### OAuth 2.0

- <mark style="background-color: #808000; color: white;">More secure authentication method</mark>
- Requires token exchange process
- Common for social media and cloud service APIs

### Handling API Responses

#### Status Codes

- **200**: Success
- **201**: Created
- **400**: Bad Request
- **401**: Unauthorized
- **404**: Not Found
- **500**: <mark style="background-color: #808000; color: white;">Server Error</mark>

#### Error Handling

```python
try:
    response = requests.get('https://api.example.com/data')
    response.raise_for_status()  # Raises exception for bad status codes
    data = response.json()
except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
```

### Best Practices

1. **Rate Limiting**: Respect API rate limits to avoid being blocked
2. **Error Handling**: Always implement proper exception handling
3. **Timeouts**: Set timeouts to prevent hanging requests
4. **Caching**: <mark style="background-color: #808000; color: white;">Cache responses when appropriate to reduce API calls</mark>
5. **Security**: Never hardcode API keys; use environment variables

### Popular Business APIs

- **Google Maps API**: Location and mapping services
- **Stripe API**: Payment processing
- **Twilio API**: SMS and communication
- **Twitter/X API**: Social media data
- **OpenAI API**: <mark style="background-color: #808000; color: white;">AI and language models</mark>

# Web Scraping

## What is Web Scraping?

Web scraping is the <mark style="background-color: #808000; color: white;">automated process of extracting data from websites</mark>. It involves fetching web pages and parsing the HTML content to extract specific information.

## When to Use Web Scraping

### Appropriate Use Cases

- **Price Monitoring**: Tracking competitor pricing
- **Market Research**: Gathering business intelligence
- **Lead Generation**: Collecting contact information
- **Content Aggregation**: Compiling information from multiple sources
- **Academic Research**: <mark style="background-color: #808000; color: white;">Collecting data for analysis</mark>

### Legal and Ethical Considerations

1. **Check robots.txt**: Respect website crawling policies
2. **Terms of Service**: Review website terms before scraping
3. **Rate Limiting**: Don't overload servers with requests
4. **Copyright**: <mark style="background-color: #808000; color: white;">Be aware of copyright and data ownership</mark>
5. **Personal Data**: Comply with privacy laws (GDPR, CCPA)

## Python Libraries for Web Scraping

### BeautifulSoup

- **Purpose**: HTML and XML parsing
- **Best For**: Static websites
- **Easy to Learn**: Great for beginners

```python
from bs4 import BeautifulSoup
import requests

# Fetch webpage
url = 'https://example.com'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Extract data
titles = soup.find_all('h2', class_='title')
for title in titles:
    print(title.text)
```

### Selenium

- **Purpose**: <mark style="background-color: #808000; color: white;">Browser automation</mark>
- **Best For**: Dynamic websites with JavaScript
- **Capabilities**: Can interact with web elements (clicking, scrolling, form filling)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Initialize driver
driver = webdriver.Chrome()
driver.get('https://example.com')

# Extract data
elements = driver.find_elements(By.CLASS_NAME, 'product-name')
for element in elements:
    print(element.text)

driver.quit()
```

### Scrapy

- **Purpose**: Full-featured web scraping framework
- **Best For**: <mark style="background-color: #808000; color: white;">Large-scale scraping projects</mark>
- **Features**: Built-in support for crawling, data extraction, and storage

### Other Useful Libraries

- **lxml**: Fast XML/HTML parser
- **Playwright**: Modern browser automation
- **httpx**: Async HTTP requests

## Web Scraping Techniques

### HTML Parsing

#### Finding Elements

```python
# By tag name
soup.find('div')
soup.find_all('p')

# By class
soup.find('div', class_='container')

# By ID
soup.find('div', id='main')

# By CSS selector
soup.select('.class-name')
soup.select('#id-name')
```

#### Navigating the DOM

- **Parent**: `element.parent`
- **Children**: `element.children`
- **Siblings**: <mark style="background-color: #808000; color: white;">`element.next_sibling`, `element.previous_sibling`</mark>

### Handling Dynamic Content

#### AJAX and JavaScript

Many modern websites load content dynamically using JavaScript. Solutions:

1. **Selenium/Playwright**: Render JavaScript in browser
2. **API Inspection**: Find underlying API calls using browser DevTools
3. **Wait Strategies**: <mark style="background-color: #808000; color: white;">Implement waits for elements to load</mark>

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Wait for element to be visible
element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, 'dynamic-content'))
)
```

### Data Extraction Patterns

#### Tables

```python
import pandas as pd

# Extract tables directly to DataFrame
tables = pd.read_html(url)
df = tables[0]  # First table on page
```

#### Lists and Pagination

```python
# Scrape multiple pages
base_url = 'https://example.com/page/'
all_data = []

for page in range(1, 11):
    url = f"{base_url}{page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Extract data from each page
    items = soup.find_all('div', class_='item')
    all_data.extend(items)
```

## Best Practices for Web Scraping

### Performance Optimization

1. **Concurrent Requests**: <mark style="background-color: #808000; color: white;">Use threading or async for parallel scraping</mark>
2. **Caching**: Store responses to avoid redundant requests
3. **Selective Parsing**: Only parse needed elements
4. **Connection Pooling**: Reuse HTTP connections

### Avoiding Detection

#### User Agents

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}
response = requests.get(url, headers=headers)
```

#### Rate Limiting

```python
import time

# Add delay between requests
time.sleep(2)  # Wait 2 seconds
```

#### Rotating Proxies

- Use proxy services to rotate IP addresses
- <mark style="background-color: #808000; color: white;">Prevents IP blocking on large-scale scraping</mark>

2. **JSON**: Structured, nested data
3. **Database**: <mark style="background-color: #808000; color: white;">SQLite, PostgreSQL, MongoDB for large datasets</mark>
4. **Excel**: Business reporting and analysis

import pandas as pd

### Error Handling and Logging

```python
import logging

# Set up logging
logging.basicConfig(level=logging.INFO)

    response = requests.get(url, timeout=10)

```

- **Solution**: Write robust selectors, implement fallback logic

### Challenge 4: Website Updates
- **Solution**: Regular monitoring and scraper maintenance
