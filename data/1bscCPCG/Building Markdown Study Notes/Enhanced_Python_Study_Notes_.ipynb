## 📊 Variables and Data Types

Variables are containers that store data. Python automatically determines the type based on the value you assign.

```python
# Basic Data Types Examples

# Numbers
age = 25                    # Integer
price = 19.99              # Float
temperature = -5.5         # Negative float

# Strings
name = "Alice"             # Text
message = 'Hello, World!'  # Single or double quotes
multiline = """This is
a multiline
string"""

# Booleans
is_student = True
has_graduated = False

# Display types
print(f"age is type: {type(age)}")
print(f"price is type: {type(price)}")
print(f"name is type: {type(name)}")
print(f"is_student is type: {type(is_student)}")
```

### 🗂️ Collection Data Types

Python provides powerful built-in data structures for organizing multiple values:

| Type | Ordered | Mutable | Duplicates | Syntax |
|------|---------|---------|------------|--------|
| **List** | ✅ | ✅ | ✅ | `[1, 2, 3]` |
| **Tuple** | ✅ | ❌ | ✅ | `(1, 2, 3)` |
| **Set** | ❌ | ✅ | ❌ | `{1, 2, 3}` |
| **Dictionary** | ✅* | ✅ | ❌ (keys) | `{"a": 1}` |

*Dictionaries maintain insertion order since Python 3.7

```python
# Collection Examples

# Lists - Ordered, mutable collections
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")  # Add item
fruits[0] = "apricot"    # Modify item
print(f"Fruits: {fruits}")

# Tuples - Ordered, immutable collections
coordinates = (10, 20)
# coordinates[0] = 15  # This would cause an error!
print(f"Coordinates: {coordinates}")

# Sets - Unordered, unique elements
unique_numbers = {1, 2, 3, 2, 1}  # Duplicates automatically removed
print(f"Unique numbers: {unique_numbers}")

# Dictionaries - Key-value pairs
student = {
    "name": "John",
    "age": 20,
    "major": "Computer Science"
}
print(f"Student: {student['name']}, Age: {student['age']}")
```

## 🔀 Control Flow

Control flow determines the order in which your code executes. Python uses **indentation** (4 spaces) to define code blocks.

```python
# If-Elif-Else Statements
score = 85

if score >= 90:
    grade = "A"
    print("🌟 Excellent!")
elif score >= 80:
    grade = "B"
    print("👍 Good job!")
elif score >= 70:
    grade = "C"
    print("👌 Satisfactory")
else:
    grade = "F"
    print("📚 Keep studying!")

print(f"Your grade: {grade}")
```

```python
# For Loops - Iterate over sequences
print("🔢 Counting:")
for i in range(1, 6):
    print(f"  Number: {i}")

print("\n🍎 Fruits:")
for fruit in ["apple", "banana", "cherry"]:
    print(f"  I like {fruit}")

# While Loops - Continue while condition is true
print("\n⏱️ Countdown:")
count = 3
while count > 0:
    print(f"  {count}...")
    count -= 1
print("  🚀 Blast off!")
```

## ⚙️ Functions

Functions are **reusable blocks of code** that perform specific tasks. They help organize code and avoid repetition.

```python
# Simple Function
def greet(name):
    """Greet someone by name"""
    return f"Hello, {name}! 👋"

# Function with default parameter
def calculate_area(length, width=1):
    """Calculate area of rectangle (default width=1 for square)"""
    return length * width

# Function with multiple return values
def get_statistics(numbers):
    """Return min, max, and average of a list"""
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

# Test functions
print(greet("Alice"))
print(f"Area of 5x3: {calculate_area(5, 3)}")
print(f"Area of 5x5: {calculate_area(5)}")

nums = [10, 20, 30, 40, 50]
minimum, maximum, average = get_statistics(nums)
print(f"Stats for {nums}:")
print(f"  Min: {minimum}, Max: {maximum}, Avg: {average}")
```

## 🏗️ Object-Oriented Programming (OOP)

OOP allows you to create custom data types called **classes** that bundle data and behavior together.

```python
class BankAccount:
    """A simple bank account class"""
    
    def __init__(self, owner, balance=0):
        """Initialize account with owner and optional starting balance"""
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        """Add money to account"""
        self.balance += amount
        print(f"💰 Deposited ${amount}. New balance: ${self.balance}")
    
    def withdraw(self, amount):
        """Remove money from account if sufficient funds"""
        if amount > self.balance:
            print("❌ Insufficient funds!")
        else:
            self.balance -= amount
            print(f"💸 Withdrew ${amount}. New balance: ${self.balance}")
    
    def get_balance(self):
        """Check current balance"""
        return f"💵 {self.owner}'s balance: ${self.balance}"

# Create and use account
account = BankAccount("Alice", 1000)
print(account.get_balance())
account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # This will fail
print(account.get_balance())
```

## 📝 List Comprehensions

A concise way to create lists based on existing sequences.

```python
# Traditional approach
squares_traditional = []
for i in range(1, 6):
    squares_traditional.append(i ** 2)
print(f"Traditional: {squares_traditional}")

# List comprehension - more concise!
squares = [i ** 2 for i in range(1, 6)]
print(f"Comprehension: {squares}")

# With condition
even_squares = [i ** 2 for i in range(1, 11) if i % 2 == 0]
print(f"Even squares: {even_squares}")

# String operations
words = ["hello", "world", "python"]
uppercase = [word.upper() for word in words]
print(f"Uppercase: {uppercase}")
```

## 🛡️ Exception Handling

Handle errors gracefully to prevent program crashes.

```python
def safe_divide(a, b):
    """Safely divide two numbers"""
    try:
        result = a / b
        print(f"✅ {a} ÷ {b} = {result}")
        return result
    except ZeroDivisionError:
        print("❌ Error: Cannot divide by zero!")
        return None
    except TypeError:
        print("❌ Error: Please provide numbers only!")
        return None
    finally:
        print("🔚 Division operation completed.\n")

# Test cases
safe_divide(10, 2)
safe_divide(10, 0)
safe_divide(10, "two")
```

**Why Visualize Data?**

## 📚 Core Visualization Libraries

### 1. Matplotlib - The Foundation 🎨

Matplotlib is Python's **most fundamental plotting library**, providing complete control over every aspect of your visualizations. It's inspired by MATLAB and can create virtually any type of static, animated, or interactive plot.

**Strengths:**
- Complete customization control
- Publication-quality output
- Wide format support (PNG, PDF, SVG)
- Works seamlessly with NumPy arrays

**Best For:** Scientific plots, detailed customization, publication graphics

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [23000, 25000, 28000, 31000, 29000, 35000]

# Create figure with multiple subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Line Plot
axes[0].plot(months, sales, marker='o', linewidth=2, color='#2196F3', markersize=8)
axes[0].set_title('📈 Monthly Sales Trend', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Month', fontsize=11)
axes[0].set_ylabel('Sales ($)', fontsize=11)
axes[0].grid(True, alpha=0.3)
axes[0].set_facecolor('#f8f9fa')

# Bar Chart
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8', '#F7DC6F']
axes[1].bar(months, sales, color=colors, edgecolor='black', linewidth=1.2)
axes[1].set_title('📊 Sales by Month', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Month', fontsize=11)
axes[1].set_ylabel('Sales ($)', fontsize=11)
axes[1].set_facecolor('#f8f9fa')

plt.tight_layout()
plt.show()

print("\n📊 Visualization Complete!")
print(f"Total Sales: ${sum(sales):,}")
print(f"Average Monthly Sales: ${sum(sales)/len(sales):,.0f}")
```

### 2. Seaborn - Statistical Beauty 📈

Built on top of Matplotlib, Seaborn provides a **high-level interface** for creating attractive statistical graphics with minimal code.

**Strengths:**
- Beautiful default themes and color palettes
- Built-in statistical functions
- Perfect for exploratory data analysis
- Seamless pandas integration

**Best For:** Statistical analysis, correlation matrices, distribution plots

```python
import seaborn as sns
import pandas as pd

# Create sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'] * 20,
    'Value': np.random.normal(100, 15, 100),
    'Group': np.random.choice(['Group 1', 'Group 2'], 100)
})

# Set style
sns.set_style("whitegrid")
sns.set_palette("husl")

# Create visualizations
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box Plot
sns.boxplot(data=data, x='Category', y='Value', ax=axes[0], palette='Set2')
axes[0].set_title('📦 Distribution by Category', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Value', fontsize=11)

# Violin Plot
sns.violinplot(data=data, x='Category', y='Value', hue='Group', ax=axes[1], split=True)
axes[1].set_title('🎻 Distribution Comparison', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Value', fontsize=11)
axes[1].legend(title='Group', loc='upper right')

plt.tight_layout()
plt.show()

print("\n📊 Statistical Summary:")
print(data.groupby('Category')['Value'].describe()[['mean', 'std', 'min', 'max']].round(2))
```

## 📊 Common Chart Types

### When to Use Each Chart Type

## 🌟 Introduction to AI Agents

An **AI Agent** is an autonomous software entity that can **perceive its environment**, **make decisions**, and **take actions** to achieve specific goals. Unlike traditional programs that follow fixed instructions, AI agents can adapt their behavior based on experience and operate independently.

**Think of AI Agents Like:**
- 🚗 **Self-driving cars** that navigate roads autonomously
- 🎮 **Video game characters** that adapt to player strategies  
- 🏠 **Smart home systems** that learn your preferences
- 💬 **Chatbots** that understand and respond to questions
- 📈 **Trading bots** that make investment decisions

### Traditional Program vs. AI Agent

| Traditional Program | AI Agent |
|-------------------|----------|
| Follows fixed rules | Learns and adapts |
| Requires explicit instructions | Makes autonomous decisions |
| Same output for same input | Behavior evolves over time |
| Cannot handle unexpected situations | Handles uncertainty |
| Deterministic | Can be probabilistic |

## 🧩 Core Components of an AI Agent

Every AI agent consists of four fundamental components that work together:

```python
# Simple AI Agent Example: Temperature Controller

class TemperatureAgent:
    """An AI agent that maintains room temperature"""
    
    def __init__(self, target_temp=22):
        self.target_temp = target_temp
        self.history = []
        print(f"🤖 Agent initialized. Target temperature: {target_temp}°C")
    
    def perceive(self, current_temp):
        """1. PERCEPTION: Sense the environment"""
        print(f"\n🌡️  Current temperature: {current_temp}°C")
        return current_temp
    
    def decide(self, current_temp):
        """2. DECISION: Determine best action"""
        difference = self.target_temp - current_temp
        
        if difference > 2:
            action = "HEAT"
            print(f"🤔 Decision: Too cold (Δ{difference}°C) → {action}")
        elif difference < -2:
            action = "COOL"
            print(f"🤔 Decision: Too hot (Δ{difference}°C) → {action}")
        else:
            action = "MAINTAIN"
            print(f"🤔 Decision: Just right (Δ{difference}°C) → {action}")
        
        return action
    
    def act(self, action):
        """3. ACTION: Execute the decision"""
        actions_map = {
            "HEAT": "🔥 Heater activated",
            "COOL": "❄️  Air conditioner activated",
            "MAINTAIN": "✅ No action needed"
        }
        print(f"⚡ Action: {actions_map[action]}")
        return action
    
    def learn(self, temp, action):
        """4. LEARNING: Remember and improve"""
        self.history.append({'temp': temp, 'action': action})
        print(f"📚 Learning: Recorded experience (total: {len(self.history)})")
    
    def run_cycle(self, current_temp):
        """Complete agent cycle: Perceive → Decide → Act → Learn"""
        print("="*50)
        temp = self.perceive(current_temp)
        decision = self.decide(temp)
        action = self.act(decision)
        self.learn(temp, action)
        return action

# Simulate the agent
agent = TemperatureAgent(target_temp=22)

# Test different temperatures
temperatures = [18, 20, 22, 25, 23]
for temp in temperatures:
    agent.run_cycle(temp)

print("\n" + "="*50)
print(f"📊 Agent Memory: {len(agent.history)} experiences recorded")
```

## 🎭 Types of AI Agents

AI agents can be classified based on their capabilities and intelligence level:

### 1. 🔵 Simple Reflex Agents

The simplest type - **reacts directly to current perceptions** without considering history.

**Characteristics:**
- No memory of past events
- Uses if-then rules
- Fast but limited

**Example:** Automatic door that opens when motion is detected

```python
class ReflexAgent:
    """Simple reflex agent - reacts to current input only"""
    
    def perceive_and_act(self, sensor_input):
        # Simple rule-based behavior
        if sensor_input == "motion_detected":
            return "🚪 OPEN door"
        elif sensor_input == "no_motion":
            return "🚪 CLOSE door"
        else:
            return "⏸️  WAIT"

# Test the reflex agent
door_agent = ReflexAgent()
print("🔵 Simple Reflex Agent Demo:")
print(f"Input: motion_detected → {door_agent.perceive_and_act('motion_detected')}")
print(f"Input: no_motion → {door_agent.perceive_and_act('no_motion')}")
```

### 2. 🟢 Model-Based Agents

**Maintains internal state** and models of how the world works.

**Characteristics:**
- Remembers past perceptions
- Has a world model
- Handles partial observability

**Example:** Robot vacuum that maps your room

```python
class VacuumAgent:
    """Model-based agent that remembers cleaned areas"""
    
    def __init__(self):
        self.cleaned_areas = set()
        self.current_position = (0, 0)
    
    def clean(self, position):
        self.current_position = position
        
        if position in self.cleaned_areas:
            return f"✅ Position {position} already clean"
        else:
            self.cleaned_areas.add(position)
            return f"🧹 Cleaning position {position}"
    
    def get_status(self):
        return f"📍 Current: {self.current_position} | Cleaned: {len(self.cleaned_areas)} areas"

# Test the model-based agent
vacuum = VacuumAgent()
print("🟢 Model-Based Agent Demo:")
print(vacuum.clean((0, 0)))
print(vacuum.clean((0, 1)))
print(vacuum.clean((0, 0)))  # Revisiting same spot
print(vacuum.get_status())
```

### 3. 🟡 Goal-Based Agents

**Works toward specific objectives** and plans sequences of actions.

**Characteristics:**
- Has explicit goals
- Plans ahead
- Evaluates future outcomes

**Example:** GPS navigation system finding the best route

```python
class NavigationAgent:
    """Goal-based agent that plans routes"""
    
    def __init__(self, start, goal):
        self.position = start
        self.goal = goal
        self.path = []
    
    def plan_route(self):
        """Simple path planning"""
        print(f"🎯 Goal: Reach {self.goal} from {self.position}")
        
        # Simple planning: move step by step
        current = list(self.position)
        while current != list(self.goal):
            if current[0] < self.goal[0]:
                current[0] += 1
            elif current[0] > self.goal[0]:
                current[0] -= 1
            elif current[1] < self.goal[1]:
                current[1] += 1
            elif current[1] > self.goal[1]:
                current[1] -= 1
            self.path.append(tuple(current))
        
        return self.path
    
    def execute(self):
        route = self.plan_route()
        print(f"📍 Planned route: {self.position}", end="")
        for step in route:
            print(f" → {step}", end="")
        print(f"\n✅ Goal reached in {len(route)} steps!")

# Test the goal-based agent
print("🟡 Goal-Based Agent Demo:")
nav_agent = NavigationAgent(start=(0, 0), goal=(3, 2))
nav_agent.execute()
```

### 4. 🔴 Learning Agents

**Improves performance through experience** using machine learning.

**Characteristics:**
- Learns from feedback
- Adapts behavior over time
- Gets better with experience

**Example:** Recommendation systems that learn your preferences

```python
class RecommendationAgent:
    """Learning agent that adapts to user preferences"""
    
    def __init__(self):
        self.preferences = {}  # Item: score
        self.interactions = 0
    
    def learn_from_feedback(self, item, liked):
        """Learn from user feedback"""
        self.interactions += 1
        
        if item not in self.preferences:
            self.preferences[item] = 0
        
        # Update preference score
        if liked:
            self.preferences[item] += 1
            feedback = "👍 Positive"
        else:
            self.preferences[item] -= 1
            feedback = "👎 Negative"
        
        print(f"{feedback} feedback for '{item}' (score: {self.preferences[item]})")
    
    def recommend(self):
        """Recommend based on learned preferences"""
        if not self.preferences:
            return "No data yet - need more interactions!"
        
        best_item = max(self.preferences, key=self.preferences.get)
        return f"🎯 Recommended: '{best_item}' (score: {self.preferences[best_item]})"
    
    def get_stats(self):
        return f"📊 Interactions: {self.interactions} | Items tracked: {len(self.preferences)}"

# Test the learning agent
print("🔴 Learning Agent Demo:\n")
rec_agent = RecommendationAgent()

# Simulate user interactions
rec_agent.learn_from_feedback("Action Movies", liked=True)
rec_agent.learn_from_feedback("Romance Movies", liked=False)
rec_agent.learn_from_feedback("Action Movies", liked=True)
rec_agent.learn_from_feedback("Comedy Movies", liked=True)
rec_agent.learn_from_feedback("Action Movies", liked=True)

print(f"\n{rec_agent.recommend()}")
print(rec_agent.get_stats())
```

## 🚀 The Future of AI Agents

AI agents are evolving rapidly with exciting developments:

**Emerging Trends:**
- 🤝 **Multi-agent systems** - Agents collaborating together
- 🧠 **Large Language Models** - More natural interaction (like ChatGPT)
- 🎯 **Few-shot learning** - Learning from minimal examples
- 🔄 **Continuous learning** - Adapting throughout lifetime
- 🦾 **Embodied AI** - Physical robots with agent capabilities

```python
# Simple Multi-Agent System Example

class TaskAgent:
    """Agent specialized in specific tasks"""
    
    def __init__(self, name, specialty):
        self.name = name
        self.specialty = specialty
    
    def can_handle(self, task):
        return task.lower() in self.specialty.lower()
    
    def execute(self, task):
        if self.can_handle(task):
            return f"✅ {self.name}: Completed '{task}'"
        else:
            return f"❌ {self.name}: Cannot handle '{task}'"

class MultiAgentSystem:
    """System coordinating multiple agents"""
    
    def __init__(self):
        self.agents = []
    
    def add_agent(self, agent):
        self.agents.append(agent)
        print(f"🤖 Added {agent.name} (specialty: {agent.specialty})")
    
    def assign_task(self, task):
        print(f"\n📋 New task: '{task}'")
        for agent in self.agents:
            if agent.can_handle(task):
                print(f"   {agent.execute(task)}")
                return
        print(f"   ⚠️  No agent available for this task")

# Create multi-agent system
print("🤖 Multi-Agent System Demo:\n")
system = MultiAgentSystem()
system.add_agent(TaskAgent("DataBot", "data analysis"))
system.add_agent(TaskAgent("CodeBot", "programming"))
system.add_agent(TaskAgent("DesignBot", "design"))

# Assign various tasks
system.assign_task("Analyze sales data")
system.assign_task("Write Python code")
system.assign_task("Create logo design")
system.assign_task("Cook dinner")  # No suitable agent
```
