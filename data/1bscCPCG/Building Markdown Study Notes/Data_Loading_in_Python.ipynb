# Data Loading in Python

This notebook shows how to load data from **CSV, Excel, SQL, and APIs**.

## 1. Loading CSV Files

```python
import pandas as pd

df = pd.read_csv("sample.csv")
print(df.head())
```

## 2. Loading Excel Files

```python
df = pd.read_excel("sample.xlsx")
print(df.head())
```

## 3. Loading Data from SQL Database

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("sample.db")
df = pd.read_sql_query("SELECT * FROM table_name", conn)

print(df.head())
conn.close()
```

## 4. Loading Data from APIs

```python
import requests

url = "https://api.github.com"
response = requests.get(url)

data = response.json()
print(data)
```
