## 2. Loops in Python

Loops are used to repeat a block of code multiple times.

### Types of Loops
- for loop
- while loop
- nested loops

### for Loop
Used when the number of iterations is known.

```python
for i in range(1, 6):
    print(i)
```

### while Loop
Used when the condition is checked before each iteration.

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

### Nested Loops
A loop inside another loop.

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

## 3. Lists in Python

Lists are used to store multiple items in a single variable.

### Creating a List

```python
numbers = [1, 2, 3, 4, 5]
print(numbers)
```

### List Operations
Common operations include append, insert, remove, and length.

```python
numbers.append(6)
numbers.remove(2)
print(numbers)
print(len(numbers))
```

### List Slicing
Slicing is used to access a portion of the list.

```python
print(numbers[1:4])
```

### List Comprehensions
A concise way to create lists.

```python
squares = [x**2 for x in range(1, 6)]
print(squares)
```

## Summary

- Python programs are easy to write and read
- Loops help automate repetitive tasks
