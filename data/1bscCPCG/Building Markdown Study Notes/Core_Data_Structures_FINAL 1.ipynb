## 1. Lists – Operations, Slicing, and Comprehensions

Lists are ordered and mutable collections used to store multiple values in a single variable.

```python
numbers = [10, 20, 30, 40, 50]
numbers.append(60)
numbers.insert(1, 15)
numbers
```

```python
numbers[1:4]
numbers[::-1]
```

```python
squares = [x**2 for x in range(1, 6)]
squares
```

## 2. Dictionaries – Key Value Pairs and Methods

Dictionaries store data in key-value pairs and allow fast lookups.

```python
student = {
    "name": "Eeshani",
    "age": 18,
    "course": "Computer Science"
}
student
```

```python
student["age"] = 19
student["college"] = "ABC University"
student
```

```python
for key, value in student.items():
    print(key, value)
```

## 3. Sets – Unique Elements and Set Operations

Sets store unique elements and support mathematical set operations.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
A.intersection(B)
```

```python
A.union(B)
```

## 4. Practice Problems with Dictionaries and Sets

```python
sentence = "python is easy and python is powerful"
word_count = {}
for word in sentence.split():
    word_count[word] = word_count.get(word, 0) + 1
word_count
```

## 5. Tuples and Other Built-in Types

Tuples are ordered but immutable collections.

```python
coordinates = (10, 20)
x, y = coordinates
x, y
```

## 6. Choosing the Right Data Structure for Business Problems

| Requirement | Data Structure | Reason |
|------------|---------------|--------|
| Ordered data | List | Index-based access |
| Unique values | Set | No duplicates |
| Records | Dictionary | Key-value mapping |
| Fixed data | Tuple | Data safety |

```python
products = {
    "Laptop": 60000,
    "Phone": 30000,
    "Tablet": 20000
}
for product, price in products.items():
    print(product, price)
```
