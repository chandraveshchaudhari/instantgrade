## Introduction to Python

Python is a **high-level, interpreted programming language** known for its simplicity and readability. Created by Guido van Rossum in 1991, Python emphasizes code readability with significant whitespace and a syntax that allows programmers to express concepts in fewer lines of code compared to languages like C++ or Java.

## 🔤 Variables and Data Types

### Basic Data Types

Python supports several fundamental data types:

**1. Numeric Types:**
- `int`: Integer numbers (e.g., 42, -17, 0)
- `float`: Floating-point numbers (e.g., 3.14, -0.001)
- `complex`: Complex numbers (e.g., 3+4j)

**2. Text Type:**
- `str`: String of characters (e.g., "Hello", 'Python')

**3. Boolean Type:**
- `bool`: True or False values

**4. None Type:**
- `NoneType`: Represents the absence of a value

Python uses **dynamic typing**, meaning you don't need to declare variable types explicitly. The interpreter determines the type at runtime based on the assigned value.

```python
# Variable assignments
age = 25              # int
height = 5.9          # float
name = "Alice"        # str
is_student = True     # bool
nothing = None        # NoneType
```

## ➕ Operators

### Arithmetic Operators
- Addition (`+`), Subtraction (`-`), Multiplication (`*`), Division (`/`)
- Floor Division (`//`), Modulus (`%`), Exponentiation (`**`)

### Comparison Operators
- Equal to (`==`), Not equal to (`!=`)
- Greater than (`>`), Less than (`<`)
- Greater than or equal to (`>=`), Less than or equal to (`<=`)

### Logical Operators
- `and`: Returns True if both conditions are true
- `or`: Returns True if at least one condition is true
- `not`: Reverses the boolean value

### Assignment Operators
- Simple assignment (`=`)
- Compound assignments (`+=`, `-=`, `*=`, `/=`, etc.)

## 🔀 Control Flow

### Conditional Statements

Python uses `if`, `elif`, and `else` for conditional execution:

```python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
```

### Loops

**For Loops:** Iterate over sequences (lists, strings, ranges)

```python
for i in range(5):
    print(i)  # Prints 0, 1, 2, 3, 4

fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
```

**While Loops:** Execute as long as a condition is true

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

**Loop Control:**
- `break`: Exits the loop immediately
- `continue`: Skips to the next iteration
- `pass`: Does nothing, used as a placeholder

## 📝 Functions

Functions are reusable blocks of code that perform specific tasks. They promote code reusability and organization.

```python
def greet(name, greeting="Hello"):
    """Function with positional and default arguments"""
    return f"{greeting}, {name}!"

# Function calls
print(greet("Alice"))              # Hello, Alice!
print(greet("Bob", "Hi"))          # Hi, Bob!
```

**Key Concepts:**
- **Parameters vs Arguments:** Parameters are variables in the function definition; arguments are values passed during the call
- **Default Arguments:** Provide default values for parameters
- **Return Values:** Functions can return single or multiple values
- **Docstrings:** Triple-quoted strings for documentation

### Lambda Functions

Anonymous, single-expression functions:

```python
square = lambda x: x ** 2
print(square(5))  # 25
```

## Introduction to Data Structures

Data structures are specialized formats for organizing, processing, and storing data. Python provides four primary built-in data structures: **lists, tuples, sets, and dictionaries**. Understanding these structures is crucial for efficient programming and algorithm implementation.

## 📋 Lists

Lists are **ordered, mutable collections** that can contain elements of different types. They are one of the most versatile data structures in Python.

### Creating and Accessing Lists

```python
# Creation
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
nested = [[1, 2], [3, 4], [5, 6]]

# Accessing elements (0-indexed)
first = numbers[0]      # 1
last = numbers[-1]      # 5 (negative indexing from end)
slice = numbers[1:4]    # [2, 3, 4]
```

### List Methods and Operations

**Adding Elements:**
- `append(item)`: Add item to end
- `insert(index, item)`: Insert at specific position
- `extend(iterable)`: Add multiple items

**Removing Elements:**
- `remove(item)`: Remove first occurrence
- `pop(index)`: Remove and return item at index (default: last)
- `clear()`: Remove all items

**Other Useful Methods:**
- `sort()`: Sort in place
- `reverse()`: Reverse in place
- `count(item)`: Count occurrences
- `index(item)`: Find index of first occurrence

```python
fruits = ['apple', 'banana', 'cherry']
fruits.append('date')           # ['apple', 'banana', 'cherry', 'date']
fruits.insert(1, 'blueberry')   # ['apple', 'blueberry', 'banana', 'cherry', 'date']
fruits.remove('banana')         # ['apple', 'blueberry', 'cherry', 'date']
```

### List Comprehensions

A concise way to create lists based on existing iterables:

```python
# Basic comprehension
squares = [x**2 for x in range(10)]

# With condition
even_squares = [x**2 for x in range(10) if x % 2 == 0]

# Nested comprehension
matrix = [[i*j for j in range(3)] for i in range(3)]
```

## 🔒 Tuples

Tuples are **ordered, immutable collections**. Once created, their elements cannot be changed, making them ideal for fixed collections of items.

```python
# Creation
coordinates = (10, 20)
single_tuple = (5,)  # Note the comma for single-element tuple
rgb_color = (255, 128, 0)

# Unpacking
x, y = coordinates
r, g, b = rgb_color

# Methods
numbers = (1, 2, 2, 3, 4, 2)
count = numbers.count(2)    # 3
index = numbers.index(3)    # 3
```

**Advantages of Tuples:**
- Faster than lists for iteration
- Can be used as dictionary keys (unlike lists)
- Protect data from accidental modification
- Useful for returning multiple values from functions

## 🎲 Sets

Sets are **unordered collections of unique elements**. They're perfect for membership testing, removing duplicates, and mathematical set operations.

```python
# Creation
fruits = {'apple', 'banana', 'cherry'}
numbers = set([1, 2, 2, 3, 3, 4])  # {1, 2, 3, 4} - duplicates removed
empty_set = set()  # Note: {} creates an empty dict, not set

# Adding and removing
fruits.add('date')
fruits.remove('banana')  # Raises error if not found
fruits.discard('grape')  # No error if not found
```

### Set Operations

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

union = set1 | set2              # {1, 2, 3, 4, 5, 6}
intersection = set1 & set2       # {3, 4}
difference = set1 - set2         # {1, 2}
symmetric_diff = set1 ^ set2     # {1, 2, 5, 6}
```

**Common Use Cases:**
- Removing duplicates from a list: `unique = list(set(my_list))`
- Fast membership testing: `if item in my_set:`
- Finding common elements between collections

## 📖 Dictionaries

Dictionaries are **unordered collections of key-value pairs**. They provide fast lookup, insertion, and deletion operations.

```python
# Creation
student = {
    'name': 'Alice',
    'age': 20,
    'grades': [85, 90, 92]
}

# Accessing values
name = student['name']           # Direct access (raises KeyError if missing)
age = student.get('age')         # Safe access (returns None if missing)
city = student.get('city', 'Unknown')  # With default value

# Adding/modifying
student['major'] = 'Computer Science'
student['age'] = 21
```

### Dictionary Methods

```python
# Keys, values, and items
keys = student.keys()       # dict_keys(['name', 'age', 'grades', 'major'])
values = student.values()   # dict_values(['Alice', 21, [85, 90, 92], 'CS'])
items = student.items()     # dict_items([('name', 'Alice'), ...])

# Iteration
for key in student:
    print(f"{key}: {student[key]}")

for key, value in student.items():
    print(f"{key}: {value}")

# Other methods
student.update({'gpa': 3.8})     # Merge dictionaries
removed = student.pop('major')   # Remove and return value
```

### Dictionary Comprehensions

```python
# Create dictionary from lists
keys = ['a', 'b', 'c']
values = [1, 2, 3]
my_dict = {k: v for k, v in zip(keys, values)}

# Filter dictionary
scores = {'Alice': 85, 'Bob': 72, 'Charlie': 90}
passed = {k: v for k, v in scores.items() if v >= 80}
```

## 🔄 Choosing the Right Data Structure

| **Structure** | **Ordered** | **Mutable** | **Duplicates** | **Best For** |
|--------------|-------------|-------------|----------------|---------------|
| List         | ✓           | ✓           | ✓              | Ordered collections, frequent modifications |
| Tuple        | ✓           | ✗           | ✓              | Fixed data, function returns, dict keys |
| Set          | ✗           | ✓           | ✗              | Unique items, membership testing |
| Dictionary   | ✗*          | ✓           | Keys: ✗        | Key-value mapping, fast lookups |

*Dictionaries maintain insertion order since Python 3.7+

## 💡 Key Takeaways

- Lists are versatile and mutable—use for general-purpose ordered collections
- Tuples provide immutability and slight performance benefits
- Sets offer O(1) membership testing and automatic duplicate removal
- Dictionaries enable efficient key-based data access
- Comprehensions provide elegant, readable ways to create collections
- Choose data structures based on your specific needs: ordering, mutability, and uniqueness requirements

## Introduction to OOP

Object-Oriented Programming is a **programming paradigm** based on the concept of "objects" that contain data (attributes) and code (methods). OOP helps organize complex programs into manageable, reusable components. Python supports OOP through classes and objects, making it ideal for building scalable applications.

## 🏗️ Classes and Objects

A **class** is a blueprint for creating objects. An **object** is an instance of a class with specific values for its attributes.

```python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis familiaris"
    
    # Constructor method
    def __init__(self, name, age):
        # Instance attributes (unique to each instance)
        self.name = name
        self.age = age
    
    # Instance method
    def bark(self):
        return f"{self.name} says Woof!"
    
    def description(self):
        return f"{self.name} is {self.age} years old"

# Creating objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

print(dog1.bark())        # Buddy says Woof!
print(dog2.description()) # Max is 5 years old
```

**Key Concepts:**
- `__init__()`: Constructor that initializes new objects
- `self`: Reference to the current instance
- **Instance attributes**: Unique to each object
- **Class attributes**: Shared across all instances

## 🔐 The Four Pillars of OOP

### 1. Encapsulation

Encapsulation is the bundling of data and methods that operate on that data within a single unit (class). It restricts direct access to some components, which is a way to prevent accidental modification.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private attribute (name mangling)
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Invalid amount"
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew ${amount}. New balance: ${self.__balance}"
        return "Insufficient funds or invalid amount"
    
    def get_balance(self):
        return self.__balance

account = BankAccount("Alice", 1000)
account.deposit(500)
# account.__balance would raise AttributeError
```

**Access Modifiers:**
- Public: `self.attribute` (accessible everywhere)
- Protected: `self._attribute` (convention, should be used within class and subclasses)
- Private: `self.__attribute` (name mangled to prevent external access)

### 2. Inheritance

Inheritance allows a class to inherit attributes and methods from another class, promoting code reuse and establishing relationships between classes.

```python
# Parent class (Base class)
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
    def make_sound(self):
        return "Some generic sound"
    
    def info(self):
        return f"{self.name} is a {self.species}"

# Child class (Derived class)
class Cat(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Cat")  # Call parent constructor
        self.breed = breed
    
    def make_sound(self):  # Method overriding
        return "Meow!"
    
    def purr(self):  # New method specific to Cat
        return f"{self.name} is purring"

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Dog")
        self.breed = breed
    
    def make_sound(self):
        return "Woof!"

# Usage
cat = Cat("Whiskers", "Persian")
dog = Dog("Buddy", "Golden Retriever")

print(cat.make_sound())  # Meow!
print(dog.make_sound())  # Woof!
print(cat.info())        # Whiskers is a Cat
```

**Types of Inheritance:**
- **Single Inheritance**: Child inherits from one parent
- **Multiple Inheritance**: Child inherits from multiple parents
- **Multilevel Inheritance**: Chain of inheritance (A → B → C)
- **Hierarchical Inheritance**: Multiple children from one parent

### 3. Polymorphism

Polymorphism allows methods to behave differently based on the object calling them. It enables a single interface to represent different underlying forms (data types).

```python
# Method overriding polymorphism
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2

# Polymorphic behavior
def print_area(shape):
    print(f"Area: {shape.area()}")

rect = Rectangle(5, 10)
circle = Circle(7)

print_area(rect)    # Area: 50
print_area(circle)  # Area: 153.93791
```

### 4. Abstraction

Abstraction hides complex implementation details and shows only essential features. Python uses abstract base classes (ABC) for this.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass
    
    @abstractmethod
    def stop_engine(self):
        pass
    
    def honk(self):  # Concrete method
        return "Beep beep!"

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
    def stop_engine(self):
        return "Car engine stopped"

# Cannot instantiate abstract class
# vehicle = Vehicle()  # Raises TypeError

car = Car()
print(car.start_engine())  # Car engine started
```

## 🛠️ Special Methods (Magic Methods)

Special methods (dunder methods) allow customization of class behavior for built-in operations.

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        # Human-readable string representation
        return f"{self.title} by {self.author}"
    
    def __repr__(self):
        # Developer-friendly representation
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    def __len__(self):
        # Length of the book
        return self.pages
    
    def __eq__(self, other):
        # Equality comparison
        return self.title == other.title and self.author == other.author

book = Book("1984", "George Orwell", 328)
print(str(book))    # 1984 by George Orwell
print(len(book))    # 328
```

**Common Magic Methods:**
- `__init__()`: Constructor
- `__str__()`: String representation for users
- `__repr__()`: String representation for developers
- `__len__()`: Define behavior for `len()`
- `__eq__()`, `__lt__()`, etc.: Comparison operators
- `__add__()`, `__sub__()`, etc.: Arithmetic operators

## 🎯 Class Methods and Static Methods

```python
class Person:
    population = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.population += 1
    
    @classmethod
    def from_birth_year(cls, name, birth_year):
        # Alternative constructor
        age = 2024 - birth_year
        return cls(name, age)
    
    @staticmethod
    def is_adult(age):
        # Utility function related to class
        return age >= 18

# Usage
person1 = Person("Alice", 25)
person2 = Person.from_birth_year("Bob", 1995)
print(Person.is_adult(16))  # False
```
