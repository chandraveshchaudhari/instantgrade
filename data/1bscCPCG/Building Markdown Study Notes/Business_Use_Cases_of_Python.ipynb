# Business Use Cases of Python

This notebook explains how Python is used in business for **Automation, Analytics, and Machine Learning**.

## 1. Automation
Automation helps businesses reduce manual work.

### Example: Renaming Multiple Files

```python
import os

folder_path = "sample_folder"

for count, filename in enumerate(os.listdir(folder_path)):
    new_name = f"file_{count}.txt"
    src = os.path.join(folder_path, filename)
    dst = os.path.join(folder_path, new_name)
    os.rename(src, dst)

print("Files renamed successfully!")
```

## 2. Business Analytics
Python helps analyze sales, marketing, and finance data.

```python
import pandas as pd

data = {
    "Product": ["A", "B", "C"],
    "Sales": [150, 200, 170]
}

df = pd.DataFrame(data)

print("Total Sales:", df["Sales"].sum())
print("Average Sales:", df["Sales"].mean())
```

## 3. Machine Learning
Businesses use ML for predictions and recommendations.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3], [4]])
y = np.array([10, 20, 30, 40])

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[5]])
print("Prediction for input 5:", prediction)
```
