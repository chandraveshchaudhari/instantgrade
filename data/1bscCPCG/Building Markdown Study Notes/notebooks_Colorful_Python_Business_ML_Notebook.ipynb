
<h2 style="color:#7C3AED;">2️⃣ Python for Machine Learning</h2>
<div style="background-color:#F5F3FF; padding:12px; border-radius:10px;">
Machine Learning helps predict future business outcomes.
</div>


```python

import numpy as np

# Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([12000, 15000, 18000, 21000, 26000])

# Convert X to matrix form
X_mean = np.mean(X)
y_mean = np.mean(y)

# Formula for slope (m)
numerator = np.sum((X - X_mean) * (y - y_mean))
denominator = np.sum((X - X_mean)**2)

m = numerator / denominator   # slope
c = y_mean - m * X_mean        # intercept

print("Slope (m):", m)
print("Intercept (c):", c)

# Prediction function
def predict(x):
    return m*x + c

# Predict for month 6
predict(6)
```
