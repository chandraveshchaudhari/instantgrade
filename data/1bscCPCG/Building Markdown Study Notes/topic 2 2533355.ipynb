## 3. Data Cleaning & Preprocessing

**Common Issues:**
- Duplicate records
- Incorrect data types
- Inconsistent values

```python
# Remove duplicates
df_clean = df_csv.drop_duplicates()

# Convert data types
# df_clean['date'] = pd.to_datetime(df_clean['date'])
```

## 4. Handling Missing Data & Outliers

**Techniques:**
- Drop
- Impute (mean, median, mode)
- Detect outliers using IQR

```python
# Missing values
df_clean.fillna(df_clean.mean(numeric_only=True), inplace=True)

# Outlier detection (IQR)
Q1 = df_clean.quantile(0.25)
Q3 = df_clean.quantile(0.75)
IQR = Q3 - Q1

df_no_outliers = df_clean[~((df_clean < (Q1 - 1.5 * IQR)) | (df_clean > (Q3 + 1.5 * IQR))).any(axis=1)]
```

## 5. Feature Engineering

**Purpose:** Convert raw data into meaningful features for models.

**Examples:**
- Encoding categories
- Scaling numerical features

```python
# One-hot encoding
df_encoded = pd.get_dummies(df_no_outliers, drop_first=True)
df_encoded.head()
```

## 6. Data Visualisation

**Tools:** Matplotlib, Seaborn

**Best Practices:**
- Keep visuals simple
- Focus on insights, not decoration

```python
import matplotlib.pyplot as plt

df_encoded.hist(figsize=(8,6))
plt.show()
```
