# 1. Core Data Structures 📊

<div style="background-color: #fff3e0; padding: 15px; border-radius: 8px; border-left: 5px solid #ff9800; margin: 20px 0;">
    <strong>💡 Key Concept:</strong> Data structures are fundamental ways of organizing and storing data in Python to enable efficient access and modification.
</div>

## 1.1 Lists 📝

Lists are **ordered, mutable sequences** that can contain elements of different types. They're one of Python's most versatile data structures.

**Key Characteristics:**
- **Mutable**: Can be modified after creation
- **Ordered**: Maintains insertion order
- **Allows duplicates**: Same value can appear multiple times
- **Heterogeneous**: Can store different data types

**Common Operations:**
- Indexing and slicing: `my_list[0]`, `my_list[1:3]`
- Appending: `my_list.append(item)`
- Extending: `my_list.extend([1, 2, 3])`
- Inserting: `my_list.insert(index, item)`
- Removing: `my_list.remove(item)`, `my_list.pop()`

**Time Complexity:**
- Access by index: O(1)
- Append: O(1)
- Insert/Delete at beginning: O(n)
- Search: O(n)

## 1.2 Tuples 🔒

Tuples are **ordered, immutable sequences** that provide data integrity and can be used as dictionary keys.

**Key Characteristics:**
- **Immutable**: Cannot be modified after creation
- **Ordered**: Maintains insertion order
- **Hashable**: Can be used as dictionary keys
- **Memory efficient**: Slightly faster than lists

**Use Cases:**
- Returning multiple values from functions
- Dictionary keys (when immutability is needed)
- Protecting data from accidental modification
- Unpacking: `x, y, z = (1, 2, 3)`

## 1.3 Dictionaries 🗂️

Dictionaries are **unordered collections of key-value pairs** that provide fast lookups.

**Key Characteristics:**
- **Key-value pairs**: Maps keys to values
- **Mutable**: Can be modified
- **Keys must be immutable**: strings, numbers, tuples
- **No duplicate keys**: Each key is unique

**Common Operations:**
- Access: `my_dict['key']` or `my_dict.get('key', default)`
- Add/Update: `my_dict['new_key'] = value`
- Remove: `del my_dict['key']`, `my_dict.pop('key')`
- Iteration: `my_dict.keys()`, `my_dict.values()`, `my_dict.items()`

**Time Complexity:**
- Access/Insert/Delete: O(1) average case
- Search: O(1) for keys, O(n) for values

## 1.4 Sets 🎯

Sets are **unordered collections of unique elements** perfect for membership testing and eliminating duplicates.

**Key Characteristics:**
- **Unordered**: No index access
- **Unique elements**: Automatically removes duplicates
- **Mutable**: Can add/remove elements
- **Mathematical operations**: Union, intersection, difference

**Common Operations:**
- Add: `my_set.add(item)`
- Remove: `my_set.remove(item)`, `my_set.discard(item)`
- Union: `set1 | set2` or `set1.union(set2)`
- Intersection: `set1 & set2` or `set1.intersection(set2)`
- Difference: `set1 - set2` or `set1.difference(set2)`

**Time Complexity:**
- Add/Remove/Membership: O(1) average case

## 1.5 Advanced Collections 🚀

**Collections Module:**

- **defaultdict**: Dictionary with default values for missing keys
- **Counter**: Dictionary subclass for counting hashable objects
- **deque**: Double-ended queue for fast appends/pops from both ends
- **namedtuple**: Tuple subclass with named fields
- **OrderedDict**: Dictionary that remembers insertion order (Python 3.7+ dicts are ordered by default)

**Choosing the Right Data Structure:**

| Need | Use |
|------|-----|
| Ordered collection with duplicates | List |
| Immutable ordered collection | Tuple |
| Key-value mapping | Dictionary |
| Unique elements only | Set |
| Fast insertions at both ends | deque |
| Count occurrences | Counter |

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 8px; border-left: 5px solid #4caf50; margin: 20px 0;">
    <strong>✅ Best Practice:</strong> Choose data structures based on your access patterns and required operations. Lists for ordered sequences, dictionaries for fast lookups, sets for uniqueness, and tuples for immutability.
</div>

```python
# Quick Examples of Core Data Structures

# Lists
fruits = ['apple', 'banana', 'cherry']
fruits.append('date')
print("List:", fruits)

# Tuples
coordinates = (10, 20, 30)
x, y, z = coordinates
print(f"Tuple unpacking: x={x}, y={y}, z={z}")

# Dictionaries
student = {'name': 'Alice', 'age': 22, 'grade': 'A'}
student['major'] = 'Computer Science'
print("Dictionary:", student)

# Sets
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print("Union:", set1 | set2)
print("Intersection:", set1 & set2)

# Collections module
from collections import Counter, defaultdict

words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
word_count = Counter(words)
print("Word count:", word_count)

# defaultdict
dd = defaultdict(list)
dd['fruits'].append('apple')
print("defaultdict:", dict(dd))
```

# 2. Object Oriented Programming (OOP) 🎓

<div style="background-color: #f3e5f5; padding: 15px; border-radius: 8px; border-left: 5px solid #9c27b0; margin: 20px 0;">
    <strong>💡 Key Concept:</strong> OOP is a programming paradigm that organizes code around objects, which contain both data (attributes) and behavior (methods).
</div>

## 2.1 The Four Pillars of OOP 🏛️

### Encapsulation 📦

Encapsulation bundles data and methods that operate on that data within a single unit (class), while restricting direct access to some components.

**Key Concepts:**
- **Public attributes**: Accessible from anywhere (no underscore)
- **Protected attributes**: Should not be accessed outside class/subclasses (single underscore `_`)
- **Private attributes**: Cannot be easily accessed outside class (double underscore `__`)
- **Properties**: Use `@property` decorator for getter/setter methods

**Benefits:**
- Data hiding and protection
- Controlled access to internal state
- Flexibility to change implementation without affecting external code
- Validation and error checking in setters

### Inheritance 👪

Inheritance allows a class to inherit attributes and methods from another class, promoting code reuse and establishing hierarchical relationships.

**Types of Inheritance:**
- **Single inheritance**: One parent class
- **Multiple inheritance**: Multiple parent classes
- **Multilevel inheritance**: Chain of inheritance
- **Hierarchical inheritance**: Multiple children from one parent

**Important Concepts:**
- `super()`: Call parent class methods
- Method overriding: Redefine parent methods in child class
- `isinstance()`: Check if object is instance of class
- `issubclass()`: Check if class is subclass of another

**Method Resolution Order (MRO):**
Python uses C3 linearization algorithm to determine method lookup order in multiple inheritance.

### Polymorphism 🎭

Polymorphism allows objects of different classes to be treated as objects of a common parent class, enabling flexible and reusable code.

**Types:**
- **Method overriding**: Subclass provides specific implementation of parent method
- **Method overloading**: Not directly supported in Python (use default arguments instead)
- **Duck typing**: "If it walks like a duck and quacks like a duck, it's a duck"
- **Operator overloading**: Define behavior for operators using magic methods

**Magic Methods:**
- `__init__`: Constructor
- `__str__`: String representation for users
- `__repr__`: String representation for developers
- `__len__`: Define length behavior
- `__add__`, `__sub__`: Arithmetic operations
- `__eq__`, `__lt__`: Comparison operations

### Abstraction 🎨

Abstraction hides complex implementation details and shows only essential features of an object.

**Implementation:**
- **Abstract Base Classes (ABC)**: Use `abc` module
- **Abstract methods**: Decorated with `@abstractmethod`
- **Interface-like classes**: Classes with only abstract methods

**Purpose:**
- Define common interface for related classes
- Force subclasses to implement specific methods
- Reduce complexity by hiding implementation details

## 2.2 Class Components 🔧

### Instance vs Class Attributes

**Instance Attributes:**
- Unique to each object instance
- Defined in `__init__` method using `self`
- Different values for different objects

**Class Attributes:**
- Shared by all instances
- Defined directly in class body
- Same value for all objects (unless overridden)

### Methods Types

**Instance Methods:**
- Take `self` as first parameter
- Can access and modify instance attributes
- Most common method type

**Class Methods:**
- Decorated with `@classmethod`
- Take `cls` as first parameter
- Can access and modify class attributes
- Often used as alternative constructors

**Static Methods:**
- Decorated with `@staticmethod`
- No `self` or `cls` parameter
- Cannot modify class or instance state
- Utility functions related to the class

## 2.3 Advanced OOP Concepts 🚀

### Composition vs Inheritance

**Composition**: "Has-a" relationship (preferred in many cases)
- Objects contain instances of other classes
- More flexible and maintainable
- Example: Car has an Engine

**Inheritance**: "Is-a" relationship
- Class extends another class
- Example: Car is a Vehicle

### Design Principles (SOLID)

**S - Single Responsibility**: Each class should have one job

**O - Open/Closed**: Open for extension, closed for modification

**L - Liskov Substitution**: Subclasses should be substitutable for parent classes

**I - Interface Segregation**: Many specific interfaces better than one general

**D - Dependency Inversion**: Depend on abstractions, not concretions

### Descriptors and Properties

Descriptors provide powerful control over attribute access using `__get__`, `__set__`, and `__delete__` methods. Properties are a simpler way to implement getters and setters.

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 8px; border-left: 5px solid #4caf50; margin: 20px 0;">
    <strong>✅ Best Practice:</strong> Favor composition over inheritance when possible. Use inheritance for "is-a" relationships and composition for "has-a" relationships. Keep classes focused on single responsibilities.
</div>

```python
# Comprehensive OOP Examples

from abc import ABC, abstractmethod

# Abstract Base Class
class Animal(ABC):
    """Abstract base class demonstrating abstraction"""
    
    species_count = 0  # Class attribute
    
    def __init__(self, name, age):
        self.name = name  # Instance attribute (public)
        self._age = age  # Protected attribute
        Animal.species_count += 1
    
    @abstractmethod
    def make_sound(self):
        """Abstract method - must be implemented by subclasses"""
        pass
    
    @property
    def age(self):
        """Getter for age"""
        return self._age
    
    @age.setter
    def age(self, value):
        """Setter with validation"""
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value
    
    @classmethod
    def get_species_count(cls):
        """Class method to access class attribute"""
        return cls.species_count
    
    @staticmethod
    def is_valid_name(name):
        """Static method - utility function"""
        return isinstance(name, str) and len(name) > 0

# Inheritance and Polymorphism
class Dog(Animal):
    """Dog class demonstrating inheritance"""
    
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # Call parent constructor
        self.breed = breed
    
    def make_sound(self):
        """Implement abstract method"""
        return f"{self.name} says: Woof!"
    
    def __str__(self):
        """String representation"""
        return f"Dog({self.name}, {self.age} years, {self.breed})"
    
    def __repr__(self):
        return f"Dog(name='{self.name}', age={self.age}, breed='{self.breed}')"

class Cat(Animal):
    """Cat class demonstrating polymorphism"""
    
    def __init__(self, name, age, indoor=True):
        super().__init__(name, age)
        self.indoor = indoor
    
    def make_sound(self):
        return f"{self.name} says: Meow!"
    
    def __str__(self):
        location = "indoor" if self.indoor else "outdoor"
        return f"Cat({self.name}, {self.age} years, {location})"

# Composition Example
class Engine:
    """Engine class for composition"""
    def __init__(self, horsepower):
        self.horsepower = horsepower
    
    def start(self):
        return "Engine started"

class Car:
    """Car has-a Engine (composition)"""
    def __init__(self, make, model, horsepower):
        self.make = make
        self.model = model
        self.engine = Engine(horsepower)  # Composition
    
    def start(self):
        return f"{self.make} {self.model}: {self.engine.start()}"

# Testing the classes
print("=== OOP Examples ===")

# Create instances
dog = Dog("Buddy", 3, "Golden Retriever")
cat = Cat("Whiskers", 2, True)

# Polymorphism in action
animals = [dog, cat]
for animal in animals:
    print(animal.make_sound())  # Different implementation for each class

# Properties
print(f"\nDog's age: {dog.age}")
dog.age = 4
print(f"Dog's new age: {dog.age}")

# Class method
print(f"\nTotal animals created: {Animal.get_species_count()}")

# Static method
print(f"Is 'Buddy' valid name? {Animal.is_valid_name('Buddy')}")

# Composition
car = Car("Toyota", "Camry", 200)
print(f"\n{car.start()}")

# String representations
print(f"\nstr: {str(dog)}")
print(f"repr: {repr(dog)}")
```

## 3.1 Performance Measurement Tools ⏱️

### Time Complexity Analysis

Understanding time complexity is crucial for writing efficient code.

**Common Time Complexities:**
- **O(1)**: Constant time - Dictionary lookup, array access by index
- **O(log n)**: Logarithmic - Binary search
- **O(n)**: Linear - Simple loops, linear search
- **O(n log n)**: Linearithmic - Efficient sorting (merge sort, quicksort)
- **O(n²)**: Quadratic - Nested loops, bubble sort
- **O(2ⁿ)**: Exponential - Recursive fibonacci (naive)

### Built-in Timing Tools

**time module:**
- `time.time()`: Wall-clock time in seconds
- `time.perf_counter()`: High-resolution timer for performance
- `time.process_time()`: CPU time (excludes sleep)

**timeit module:**
- More accurate for benchmarking small code snippets
- Automatically runs code multiple times
- Disables garbage collection for consistent results
- Usage: `timeit.timeit(stmt, setup, number=1000000)`

**IPython magic commands:**
- `%time`: Time a single execution
- `%timeit`: Time multiple executions (average)
- `%%timeit`: Time entire cell
- `%prun`: Profile function calls

### Profiling Tools 🔍

**cProfile:**
- Built-in deterministic profiler
- Shows function call counts and time spent
- Usage: `cProfile.run('function()')`

**line_profiler:**
- Line-by-line profiling
- Shows time spent on each line
- Install: `pip install line_profiler`

**memory_profiler:**
- Monitor memory usage line-by-line
- Detect memory leaks
- Install: `pip install memory_profiler`

## 3.2 Memory Optimization 💾

### Memory Measurement

**sys.getsizeof():**
- Returns size of object in bytes
- Doesn't include size of objects referenced

**tracemalloc module:**
- Track memory allocations
- Find memory leaks
- Compare snapshots

### Memory-Efficient Techniques

**Generators vs Lists:**
- Generators: Lazy evaluation, memory efficient
- Use for large datasets or infinite sequences
- Syntax: `(x for x in range(1000000))`

**__slots__:**
- Reduce memory overhead for classes
- Define fixed set of attributes
- Prevents `__dict__` creation

**Appropriate Data Structures:**
- Use `array` for numeric data instead of lists
- Use `set` for membership testing
- Use `collections.deque` for queues

## 3.5 Best Practices 🎯

### Performance Optimization

1. **Measure before optimizing**: Profile to find real bottlenecks
2. **Use appropriate data structures**: Right tool for the job
3. **Avoid premature optimization**: Readability first, optimize when needed
4. **Cache expensive computations**: `functools.lru_cache`
5. **Vectorize with NumPy**: Avoid Python loops for numerical data

### Visualization Best Practices

1. **Choose appropriate plot type**: Match visualization to data
2. **Label everything**: Axes, titles, legends
3. **Use color purposefully**: Highlight important information
4. **Keep it simple**: Avoid chart junk
5. **Consider audience**: Technical vs general audience
6. **Make it accessible**: Color-blind friendly palettes

<div style="background-color: #e8f5e9; padding: 15px; border-radius: 8px; border-left: 5px solid #4caf50; margin: 20px 0;">

</div>

```python
# Performance Measurement Examples

import time
import timeit
from functools import lru_cache
import sys

print("=== Performance Measurement Examples ===")

# 1. Time measurement
def slow_function():
    """Simulate a slow operation"""
    total = 0
    for i in range(1000000):
        total += i
    return total

# Using time module
start = time.perf_counter()
result = slow_function()
end = time.perf_counter()
print(f"\nExecution time: {end - start:.4f} seconds")

# 2. Comparing data structures performance
list_time = timeit.timeit('1000000 in my_list', setup='my_list = list(range(1000000))', number=100)
set_time = timeit.timeit('1000000 in my_set', setup='my_set = set(range(1000000))', number=100)

print(f"\nList membership test: {list_time:.6f} seconds")
print(f"Set membership test: {set_time:.6f} seconds")
print(f"Set is {list_time/set_time:.1f}x faster")

# 3. Memory usage
list_example = [i for i in range(1000)]
generator_example = (i for i in range(1000))

print(f"\nList size: {sys.getsizeof(list_example)} bytes")
print(f"Generator size: {sys.getsizeof(generator_example)} bytes")

# 4. Caching demonstration
def fibonacci_slow(n):
    """Fibonacci without caching - O(2^n)"""
    if n < 2:
        return n
    return fibonacci_slow(n-1) + fibonacci_slow(n-2)

@lru_cache(maxsize=None)
def fibonacci_fast(n):
    """Fibonacci with caching - O(n)"""
    if n < 2:
        return n
    return fibonacci_fast(n-1) + fibonacci_fast(n-2)

# Compare caching impact
n = 30
start = time.perf_counter()
result_slow = fibonacci_slow(n)
time_slow = time.perf_counter() - start

start = time.perf_counter()
result_fast = fibonacci_fast(n)
time_fast = time.perf_counter() - start

print(f"\nFibonacci({n}) without cache: {time_slow:.4f} seconds")
print(f"Fibonacci({n}) with cache: {time_fast:.6f} seconds")
print(f"Speedup: {time_slow/time_fast:.0f}x faster")
```

```python
# Visualization Examples

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 4)

print("=== Visualization Examples ===")

# 1. Performance comparison visualization
sizes = [10, 100, 1000, 10000, 100000]
list_times = []
set_times = []

for size in sizes:
    list_time = timeit.timeit(
        f'{size-1} in my_list',
        setup=f'my_list = list(range({size}))',
        number=1000
    )
    set_time = timeit.timeit(
        f'{size-1} in my_set',
        setup=f'my_set = set(range({size}))',
        number=1000
    )
    list_times.append(list_time)
    set_times.append(set_time)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1: Line plot comparison
axes[0].plot(sizes, list_times, marker='o', label='List', linewidth=2)
axes[0].plot(sizes, set_times, marker='s', label='Set', linewidth=2)
axes[0].set_xlabel('Data Structure Size', fontsize=12)
axes[0].set_ylabel('Time (seconds)', fontsize=12)
axes[0].set_title('List vs Set Performance', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].set_xscale('log')
axes[0].set_yscale('log')
axes[0].grid(True, alpha=0.3)

# Plot 2: Bar chart of algorithm complexities
algorithms = ['O(1)', 'O(log n)', 'O(n)', 'O(n log n)', 'O(n²)']
n = 1000
complexities = [1, np.log2(n), n, n*np.log2(n), n**2]

colors = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c', '#c0392b']
axes[1].bar(algorithms, complexities, color=colors, alpha=0.7)
axes[1].set_ylabel('Operations (n=1000)', fontsize=12)
axes[1].set_title('Time Complexity Comparison', fontsize=14, fontweight='bold')
axes[1].set_yscale('log')
axes[1].grid(True, alpha=0.3, axis='y')

# Plot 3: Memory usage comparison
data_sizes = [100, 1000, 10000, 100000]
list_memory = [sys.getsizeof([i for i in range(size)]) for size in data_sizes]
tuple_memory = [sys.getsizeof(tuple(i for i in range(size))) for size in data_sizes]
set_memory = [sys.getsizeof({i for i in range(size)}) for size in data_sizes]

x = np.arange(len(data_sizes))
width = 0.25

axes[2].bar(x - width, list_memory, width, label='List', alpha=0.8)
axes[2].bar(x, tuple_memory, width, label='Tuple', alpha=0.8)
axes[2].bar(x + width, set_memory, width, label='Set', alpha=0.8)
axes[2].set_xlabel('Data Size', fontsize=12)
axes[2].set_ylabel('Memory (bytes)', fontsize=12)
axes[2].set_title('Memory Usage by Data Structure', fontsize=14, fontweight='bold')
axes[2].set_xticks(x)
axes[2].set_xticklabels(data_sizes)
axes[2].legend()
axes[2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\n✅ Visualizations created successfully!")
```

    <ul style="line-height: 1.8;">
        <li>Master the four pillars: Encapsulation, Inheritance, Polymorphism, Abstraction</li>
        <li>Favor composition over inheritance for flexibility</li>
        <li>Follow SOLID principles for maintainable code</li>
    </ul>
