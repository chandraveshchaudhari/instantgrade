## 3. Implementing ML Algorithms from Scratch

**Algorithms Covered:**
- Linear Regression
- KNN
- Decision Trees (basic)

**Why From Scratch?**
- Understand math + logic
- Avoid black-box learning

```python
# Simple Linear Regression from scratch
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

mean_x = sum(x)/len(x)
mean_y = sum(y)/len(y)

num = sum((x[i]-mean_x)*(y[i]-mean_y) for i in range(len(x)))
den = sum((x[i]-mean_x)**2 for i in range(len(x)))

b1 = num/den
b0 = mean_y - b1*mean_x

print("y =", b0, "+", b1, "x")
```
