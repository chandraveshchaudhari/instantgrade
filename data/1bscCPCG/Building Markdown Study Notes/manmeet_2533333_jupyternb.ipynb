## 1. Lists: Operations, Slicing, and Comprehensions

### List Operations
- `append(x)` – adds element to end
- `insert(i, x)` – inserts at index
- `remove(x)` – removes first occurrence
- `pop()` – removes and returns last element
- `sort()` – sorts list
- `reverse()` – reverses list


```python
nums = [1, 2, 3]
nums.append(4)
nums.pop()
nums
```

### List Slicing
- Syntax: `list[start:end:step]`
- Does not modify original list


```python
arr = [10, 20, 30, 40, 50]
arr[1:4]
```

### List Comprehensions
- Compact way to create lists
- Syntax: `[expression for item in iterable if condition]`


```python
squares = [x**2 for x in range(1, 6)]
squares
```

## 2. Arrays, Linked Lists, and Stacks

### Arrays
- Stored in contiguous memory
- Fast access using index (O(1))
- Fixed size (in low-level languages)


### Linked Lists
- Elements stored as nodes
- Each node has data + reference
- Types: Singly, Doubly, Circular
- Insertion/Deletion is efficient


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

### Stacks
- Follows LIFO (Last In First Out)
- Operations: push, pop, peek
- Can be implemented using lists


```python
stack = []
stack.append(10)  # push
stack.append(20)
stack.pop()      # pop
```

### Time Complexity of List Operations

| Operation | Time Complexity |
|----------|----------------|
| Indexing | O(1) |
| Append   | O(1) |
| Insert   | O(n) |
| Delete   | O(n) |
| Search   | O(n) |

👉 Important for interviews and competitive programming.

### Common List Comprehension Patterns
- Filtering values
- Nested loops
- Conditional expressions


```python
# even numbers
evens = [x for x in range(1, 21) if x % 2 == 0]

# nested comprehension
pairs = [(i, j) for i in range(2) for j in range(3)]
evens, pairs
```

## Comparison: Arrays vs Linked Lists

| Feature | Array | Linked List |
|--------|-------|-------------|
| Memory | Contiguous | Non-contiguous |
| Access | Fast (O(1)) | Slow (O(n)) |
| Insert/Delete | Costly | Efficient |
| Cache Friendly | Yes | No |
