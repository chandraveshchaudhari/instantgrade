## 2.2 Variables

Variables are containers for storing data values.

### Variable Rules:
- Must start with letter or underscore
- Can contain letters, numbers, underscores
- Case-sensitive (`age` ≠ `Age`)
- Cannot use Python keywords (if, for, while, etc.)

```python
# Valid variable names
name = "Alice"
age = 25
_private = "hidden"
user_name = "alice123"
MAX_VALUE = 100

# Invalid variable names (commented to avoid errors)
# 2name = "Bob"        # Can't start with number
# user-name = "test"   # Can't use hyphens
# for = 5              # Can't use keywords

print("Name:", name)
print("Age:", age)
```

### Variable Assignment

```python
# Single assignment
x = 10

# Multiple assignment
a, b, c = 1, 2, 3
print(f"a={a}, b={b}, c={c}")

# Same value to multiple variables
x = y = z = 0
print(f"x={x}, y={y}, z={z}")

# Swapping variables
p, q = 5, 10
print(f"Before swap: p={p}, q={q}")
p, q = q, p
print(f"After swap: p={p}, q={q}")
```

## 2.3 Data Types

Python has several built-in data types.

```python
# Integer (int) - whole numbers
age = 25
temperature = -5
print(f"age = {age}, type = {type(age)}")

# Float - decimal numbers
price = 19.99
pi = 3.14159
print(f"price = {price}, type = {type(price)}")

# String (str) - text
name = "Alice"
message = 'Hello World'
print(f"name = {name}, type = {type(name)}")

# Boolean (bool) - True or False
is_student = True
is_working = False
print(f"is_student = {is_student}, type = {type(is_student)}")

# NoneType - represents absence of value
result = None
print(f"result = {result}, type = {type(result)}")
```

## 2.4 Type Conversion (Casting)

Converting between different data types:

```python
# String to integer
age_str = "25"
age_int = int(age_str)
print(f"String '{age_str}' → int: {age_int}")

# String to float
price_str = "19.99"
price_float = float(price_str)
print(f"String '{price_str}' → float: {price_float}")

# Integer/Float to string
number = 42
number_str = str(number)
print(f"Int {number} → string: '{number_str}'")

# Float to integer (truncates decimal)
price = 19.99
price_int = int(price)
print(f"Float {price} → int: {price_int} (decimal removed)")

# Integer to boolean
print(f"bool(0) = {bool(0)}")    # False
print(f"bool(1) = {bool(1)}")    # True
print(f"bool(42) = {bool(42)}")  # True
```

## 2.5 String Operations

### String Creation and Basic Operations

```python
# Different ways to create strings
single_quote = 'Hello'
double_quote = "World"
triple_quote = '''This is a
multi-line
string'''

# String concatenation (joining)
greeting = single_quote + " " + double_quote
print("Concatenation:", greeting)

# String repetition
laugh = "ha" * 3
print("Repetition:", laugh)

# String length
text = "Python"
print(f"Length of '{text}':", len(text))
```

### String Indexing and Slicing

Strings are indexed starting from 0.

```python
text = "Python Programming"

# Indexing (accessing single character)
print("First character:", text[0])     # P
print("Last character:", text[-1])     # g
print("Third character:", text[2])     # t

# Slicing [start:end:step]
print("First 6 chars:", text[0:6])     # Python
print("From index 7:", text[7:])       # Programming
print("Last 4 chars:", text[-4:])      # ming
print("Every 2nd char:", text[::2])    # Pto rgamn
print("Reverse string:", text[::-1])   # gnimmargorP nohtyP
```

### String Formatting

Multiple ways to format strings with variables:

```python
name = "Bob"
age = 30
height = 5.9

# Method 1: f-strings (Python 3.6+) - RECOMMENDED
message1 = f"My name is {name}, I am {age} years old and {height} feet tall."
print("f-string:", message1)

# Method 2: format() method
message2 = "My name is {}, I am {} years old.".format(name, age)
print("format():", message2)

# Method 3: % formatting (old style)
message3 = "My name is %s, I am %d years old." % (name, age)
print("% format:", message3)

# Method 4: Simple concatenation
message4 = "My name is " + name + ", I am " + str(age) + " years old."
print("Concatenation:", message4)
```

### Common String Methods

```python
text = "  Python Programming  "

print(f"Original: '{text}'")
print(f"Upper: {text.upper()}")
print(f"Lower: {text.lower()}")
print(f"Title Case: {text.title()}")
print(f"Strip whitespace: '{text.strip()}'")
print(f"Replace: {text.replace('Python', 'Java')}")
print(f"Split: {text.strip().split()}")
print(f"Count 'o': {text.count('o')}")
print(f"Find 'gram': {text.find('gram')}")
print(f"Starts with 'Python': {text.strip().startswith('Python')}")
print(f"Is digit: {'123'.isdigit()}")
print(f"Is alpha: {'abc'.isalpha()}")
```

## 2.6 Basic Control Flow

### If-Elif-Else Statements

```python
# Simple if statement
age = 18
if age >= 18:
    print("You are an adult")

# If-else statement
temperature = 25
if temperature > 30:
    print("It's hot!")
else:
    print("It's comfortable")

# If-elif-else statement
score = 75
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Score: {score}, Grade: {grade}")
```

### Comparison Operators

```python
x = 10
y = 5

print(f"{x} == {y}:", x == y)   # Equal to
print(f"{x} != {y}:", x != y)   # Not equal to
print(f"{x} > {y}:", x > y)     # Greater than
print(f"{x} < {y}:", x < y)     # Less than
print(f"{x} >= {y}:", x >= y)   # Greater than or equal
print(f"{x} <= {y}:", x <= y)   # Less than or equal
```

### Logical Operators

```python
age = 25
has_license = True

# AND operator - both conditions must be True
if age >= 18 and has_license:
    print("Can drive")

# OR operator - at least one condition must be True
is_weekend = True
is_holiday = False
if is_weekend or is_holiday:
    print("Time to relax!")

# NOT operator - reverses the boolean value
is_raining = False
if not is_raining:
    print("No umbrella needed")
```

### Variable Scope

Scope determines where a variable can be accessed.

```python
# Global variable (accessible everywhere)
global_var = "I am global"

def my_function():
    # Local variable (only accessible inside function)
    local_var = "I am local"
    print("Inside function:", global_var)
    print("Inside function:", local_var)

my_function()
print("Outside function:", global_var)
# print(local_var)  # This would cause an error!
```

## 3.2 Numeric Data Types and Operations

### Integer Operations

```python
# Basic arithmetic operators
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")            # Returns float
print(f"Floor Division: {a} // {b} = {a // b}")    # Returns int
print(f"Modulus (remainder): {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
```

### Float Operations

```python
# Float arithmetic
x = 10.5
y = 2.5

print(f"Addition: {x} + {y} = {x + y}")
print(f"Subtraction: {x} - {y} = {x - y}")
print(f"Multiplication: {x} * {y} = {x * y}")
print(f"Division: {x} / {y} = {x / y}")

# Rounding floats
pi = 3.14159265359
print(f"Original: {pi}")
print(f"Rounded to 2 decimals: {round(pi, 2)}")
print(f"Rounded to nearest int: {round(pi)}")
```

### Compound Assignment Operators

Shorthand for updating variables:

```python
# Compound assignment operators
count = 10
print(f"Initial value: {count}")

count += 5    # Same as: count = count + 5
print(f"After += 5: {count}")

count -= 3    # Same as: count = count - 3
print(f"After -= 3: {count}")

count *= 2    # Same as: count = count * 2
print(f"After *= 2: {count}")

count /= 4    # Same as: count = count / 4
print(f"After /= 4: {count}")

count //= 2   # Same as: count = count // 2
print(f"After //= 2: {count}")

count %= 3    # Same as: count = count % 3
print(f"After %= 3: {count}")

count **= 3   # Same as: count = count ** 3
print(f"After **= 3: {count}")
```

### Mathematical Functions

Python's `math` module provides many useful functions:

```python
import math

# Common math functions
print(f"Square root of 16: {math.sqrt(16)}")
print(f"Power (2^3): {math.pow(2, 3)}")
print(f"Absolute value of -10: {abs(-10)}")
print(f"Floor of 4.7: {math.floor(4.7)}")
print(f"Ceiling of 4.1: {math.ceil(4.1)}")

# Trigonometric functions (angles in radians)
angle = math.pi / 4  # 45 degrees
print(f"sin(45°): {math.sin(angle)}")
print(f"cos(45°): {math.cos(angle)}")

# Constants
print(f"Pi: {math.pi}")
print(f"e: {math.e}")
```

## 3.3 String Operators

### Concatenation and Repetition

```python
# Concatenation with +
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print("Full name:", full_name)

# Repetition with *
separator = "=" * 40
print(separator)
print("Title")
print(separator)

# Cannot directly concatenate string and number
age = 25
# message = "I am " + age + " years old"  # ERROR!
message = "I am " + str(age) + " years old"  # CORRECT
print(message)

# Better way: f-strings
message = f"I am {age} years old"
print(message)
```

### Membership Operators

Check if a substring exists in a string:

```python
text = "Python Programming"

# 'in' operator
print("'Python' in text:", "Python" in text)        # True
print("'Java' in text:", "Java" in text)            # False

# 'not in' operator
print("'Java' not in text:", "Java" not in text)    # True

# Case-sensitive
print("'python' in text:", "python" in text)        # False
print("'python' in text.lower():", "python" in text.lower())  # True
```

## 3.4 Boolean Operators

### Logical Operators

```python
# AND operator (both must be True)
print("True and True:", True and True)      # True
print("True and False:", True and False)    # False
print("False and False:", False and False)  # False

# OR operator (at least one must be True)
print("True or False:", True or False)      # True
print("False or False:", False or False)    # False

# NOT operator (reverses the value)
print("not True:", not True)                # False
print("not False:", not False)              # True
```

### Comparison Chains

```python
# Python allows chaining comparisons
x = 5

# Traditional way
print("Traditional:", x > 0 and x < 10)

# Python way (cleaner)
print("Chained:", 0 < x < 10)

# More examples
age = 25
print("18 <= age <= 65:", 18 <= age <= 65)  # True

score = 85
print("80 <= score < 90:", 80 <= score < 90)  # True
```

### Identity Operators

Check if two variables refer to the same object in memory:

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

# 'is' operator (checks identity, not value)
print("a is b:", a is b)        # False (different objects)
print("a is c:", a is c)        # True (same object)

# '==' operator (checks value)
print("a == b:", a == b)        # True (same values)

# 'is not' operator
print("a is not b:", a is not b)  # True

# Common use: checking for None
result = None
if result is None:
    print("Result is None")
```

## 3.5 Operator Precedence

Order in which operations are performed:

```python
# Operator precedence (from highest to lowest):
# 1. Parentheses ()
# 2. Exponentiation **
# 3. Unary +, -, not
# 4. *, /, //, %
# 5. +, -
# 6. Comparison operators
# 7. and
# 8. or

# Examples
result1 = 2 + 3 * 4
print("2 + 3 * 4 =", result1)  # 14 (multiplication first)

result2 = (2 + 3) * 4
print("(2 + 3) * 4 =", result2)  # 20 (parentheses first)

result3 = 10 / 2 * 3
print("10 / 2 * 3 =", result3)  # 15.0 (left to right)

result4 = 2 ** 3 ** 2
print("2 ** 3 ** 2 =", result4)  # 512 (right to left for **)

# Complex expression
result5 = 5 + 3 * 2 ** 2 - 1
print("5 + 3 * 2 ** 2 - 1 =", result5)  # 16
# Step by step: 5 + 3 * 4 - 1 → 5 + 12 - 1 → 16
```

## 3.6 Practical Examples

### Example 1: Temperature Converter

```python
# Celsius to Fahrenheit converter
print("=== Temperature Converter ===")

celsius = float(input("Enter temperature in Celsius: "))

# Formula: F = (C × 9/5) + 32
fahrenheit = (celsius * 9/5) + 32

print(f"{celsius}°C = {fahrenheit}°F")

# Bonus: Convert back
celsius_back = (fahrenheit - 32) * 5/9
print(f"Verification: {fahrenheit}°F = {celsius_back}°C")
```

### Example 2: BMI Calculator

```python
# BMI = weight (kg) / height^2 (m)
print("=== BMI Calculator ===")

weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi = weight / (height ** 2)

print(f"\nYour BMI: {bmi:.2f}")

# Classify BMI
if bmi < 18.5:
    category = "Underweight"
elif 18.5 <= bmi < 25:
    category = "Normal weight"
elif 25 <= bmi < 30:
    category = "Overweight"
else:
    category = "Obese"

print(f"Category: {category}")
```

### Example 3: Shopping Cart Calculator

```python
# Shopping cart with discount
print("=== Shopping Cart ===")

item1_price = float(input("Enter price of item 1: "))
item2_price = float(input("Enter price of item 2: "))
item3_price = float(input("Enter price of item 3: "))

# Calculate subtotal
subtotal = item1_price + item2_price + item3_price
print(f"\nSubtotal: ${subtotal:.2f}")

# Apply discount if subtotal > 100
if subtotal > 100:
    discount = subtotal * 0.10  # 10% discount
    print(f"Discount (10%): -${discount:.2f}")
else:
    discount = 0
    print("No discount applied (spend > $100 for 10% off)")

# Calculate tax
tax_rate = 0.08  # 8% tax
tax = (subtotal - discount) * tax_rate
print(f"Tax (8%): ${tax:.2f}")

# Calculate total
total = subtotal - discount + tax
print(f"\nTotal: ${total:.2f}")
```

## 3.7 Common Pitfalls and Tips

### Common Mistakes to Avoid

```python
# 1. Division by zero
# x = 10 / 0  # ERROR: ZeroDivisionError

# Safe division
numerator = 10
denominator = 0
if denominator != 0:
    result = numerator / denominator
else:
    print("Cannot divide by zero!")

# 2. Type mismatch
# age = input("Enter age: ")  # Returns string
# next_year = age + 1  # ERROR: can't add int to string

# Correct way
age = int(input("Enter age: "))
next_year = age + 1
print(f"Next year you will be {next_year}")

# 3. Comparing floats
a = 0.1 + 0.2
b = 0.3
print(f"a = {a}")
print(f"a == b: {a == b}")  # May be False due to floating point precision

# Better way: use round() or math.isclose()
print(f"round(a, 1) == b: {round(a, 1) == b}")

# 4. Modifying strings
text = "Hello"
# text[0] = "h"  # ERROR: strings are immutable

# Correct way: create new string
text = "h" + text[1:]
print(text)
```
