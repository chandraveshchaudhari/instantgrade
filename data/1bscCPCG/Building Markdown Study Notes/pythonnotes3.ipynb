## 1. List Comprehensions and Generator Expressions

### List Comprehension

```python
# Traditional way
squares = []
for i in range(5):
    squares.append(i**2)

print(squares)

# List comprehension
squares_comp = [i**2 for i in range(5)]
print(squares_comp)
```

### Generator Expression

```python
gen = (i**2 for i in range(5))

for value in gen:
    print(value)
```

## 2. File Input and Output (CSV, Excel, JSON, XML)

### Text File Example

```python
# Writing to a file
with open("sample.txt", "w") as f:
    f.write("Hello, this is a sample file.")

# Reading from a file
with open("sample.txt", "r") as f:
    content = f.read()
    print(content)
```

### CSV File Example

```python
import csv

# Writing CSV
with open("data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Gunjan", 20])

# Reading CSV
with open("data.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

### JSON Example

```python
import json

data = {"name": "Gunjan", "age": 20}

# Write JSON
with open("data.json", "w") as f:
    json.dump(data, f)

# Read JSON
with open("data.json", "r") as f:
    loaded = json.load(f)
    print(loaded)
```

## 3. Error Handling and Exceptions

```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
    print(result)
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
finally:
    print("Execution complete.")
```

## 4. Working with Libraries (NumPy, Pandas, Matplotlib)

### NumPy Example

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print("Array:", arr)
print("Mean:", np.mean(arr))
```

### Pandas Example

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Sales": [100, 200, 150]
}

df = pd.DataFrame(data)
print(df)
print("Total Sales:", df["Sales"].sum())
```

### Matplotlib Example

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker="o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Chart")
plt.show()
```

## 5. Business Data Formats (Excel, PDFs, APIs)

### Excel Example with Pandas
