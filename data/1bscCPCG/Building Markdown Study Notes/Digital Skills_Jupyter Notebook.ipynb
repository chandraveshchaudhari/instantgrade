Intermediate Python focuses on writing efficient, readable, and reusable code.

## List Comprehensions and Generator Expressions

List comprehensions provide a compact syntax to create lists using a single line of code. 
They improve readability and often perform faster than traditional loops. 
Generator expressions are similar but produce values lazily, meaning they generate 
items one at a time instead of storing the entire sequence in memory. 
This makes generators highly memory‑efficient for large datasets.


```
# List comprehension
squares = [x**2 for x in range(10)]
print(squares)

# Generator expression
gen = (x**2 for x in range(10))
for g in gen:
    print(g)
```

## File Input and Output (I/O)

File handling allows programs to store and retrieve data permanently. 
Python provides built‑in functions to read and write text files, CSV files, JSON files, 
and other structured formats. The `with` statement is recommended because it 
automatically closes the file after use, preventing resource leaks.


```
with open('example.txt','w') as f:
    f.write('Hello Python')

with open('example.txt','r') as f:
    print(f.read())
```

## Error Handling and Exceptions

Programs often encounter unexpected situations such as invalid input, missing files, 
or mathematical errors. Exception handling ensures that the program does not crash 
and instead responds gracefully. Python uses `try`, `except`, `else`, and `finally` 
blocks to manage runtime errors effectively.


```
try:
    num = int(input("Enter number: "))
    print(10/num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

## Bubble Sort Algorithm

Sorting algorithms arrange data in a specific order such as ascending or descending. 
Bubble sort repeatedly compares adjacent elements and swaps them if they are in the 
wrong order. Although simple to understand, it is not efficient for large datasets 
because its time complexity is O(n²).


```
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(bubble_sort([5,3,8,1,2]))
```

## Stack Data Structure

A stack is a linear data structure that follows the Last‑In‑First‑Out (LIFO) principle. 
Operations such as `push` (insert) and `pop` (remove) are used to manage stack elements. 
Stacks are widely used in expression evaluation, undo operations, and recursion handling.


```
stack = []
stack.append(10)
stack.append(20)
print(stack.pop())
print(stack)
```

# Topic 3: Classification Models and Support Vector Machines

Classification models are supervised learning techniques used to predict categorical 
outcomes such as spam detection, disease diagnosis, or customer churn prediction. 
These models learn patterns from labelled datasets and then classify new observations.


## Logistic Regression

Logistic regression is a statistical model used for binary or multi‑class classification. 
It estimates probabilities using the logistic (sigmoid) function and predicts the class 
with the highest probability.


```
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
model = LogisticRegression(max_iter=200)
model.fit(X, y)
print(model.score(X, y))
```

## Naive Bayes

Naive Bayes classifiers are probabilistic models based on Bayes’ theorem. 
They assume that features are independent of each other, which simplifies computation 
and allows fast predictions even for large datasets. 
They are commonly used in text classification and spam filtering.


```
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y)
print(model.score(X, y))
```

## Support Vector Machines (SVM)

Support Vector Machines are powerful classification algorithms that find the optimal 
hyperplane separating different classes. Kernel functions allow SVMs to handle 
non‑linear decision boundaries, making them effective for complex datasets such as 
image recognition and sentiment analysis.


```
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X, y)
print(svm_model.score(X, y))
```
