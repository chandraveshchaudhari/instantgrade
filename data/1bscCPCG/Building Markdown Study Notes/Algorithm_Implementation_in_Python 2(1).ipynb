## 7.1 Understanding Algorithms and Data Structures

**Algorithm:** A finite, well-defined sequence of steps used to solve a problem.

**Data Structure:** A method of organizing and storing data to enable efficient access and modification.

**Importance of Algorithms in Problem Solving:**
- Improve efficiency and accuracy
- Enable scalable solutions

```python
# Linear Search Example
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

data = [4, 7, 1, 9, 3]
print(linear_search(data, 9))
```

## 7.4 Implementing Sorting Algorithms

**Sorting:** Process of arranging data in a particular order.

**Bubble Sort:**
- Repeatedly swaps adjacent elements if they are in the wrong order

```python
# Bubble Sort Implementation
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

print(bubble_sort([5, 1, 4, 2, 8]))
```

## 7.5 Advanced Sorting Techniques

**Need for Advanced Sorting:**
- Handle large datasets efficiently

**Merge Sort:** Divide-and-conquer sorting algorithm

```python
# Merge Sort Implementation
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

print(merge_sort([3, 6, 1, 5, 2, 4]))
```

## 7.6 LeetCode Easy Problems (Conceptual)

- Test basic understanding of arrays, strings, and loops

**Example: Two Sum Problem**

```python
# Two Sum Solution
def two_sum(nums, target):
    lookup = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in lookup:
            return [lookup[diff], i]
        lookup[num] = i

print(two_sum([2, 7, 11, 15], 9))
```

```python
# Non-optimized vs Optimized Example
def sum_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

def sum_n_optimized(n):
    return n * (n + 1) // 2
```

## 7.11 Dynamic Programming and Greedy Algorithms

**Dynamic Programming:** Solves problems by breaking them into overlapping subproblems.

**Greedy Algorithms:** Make locally optimal choices at each step.

```python
# Fibonacci using Dynamic Programming
def fibonacci(n):
    dp = [0, 1]
    for i in range(2, n + 1):
        dp.append(dp[i - 1] + dp[i - 2])
    return dp[n]

print(fibonacci(10))
```
