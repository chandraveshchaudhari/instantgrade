## 1. Writing Clean and Modular Code

```python
# Example: modular function design

def calculate_total(price, tax_rate):
    return price + (price * tax_rate)

calculate_total(100, 0.18)
```

## 2. Design Patterns for ML Applications

```python
# Example: Factory pattern (simplified)

class ModelFactory:
    def get_model(self, model_type):
        if model_type == 'linear':
            return 'Linear Regression Model'
        elif model_type == 'tree':
            return 'Decision Tree Model'

factory = ModelFactory()
factory.get_model('linear')
```

## 3. Documentation Best Practices

```python
def add(a, b):
    """
    Adds two numbers.

    Args:
        a (int): first number
        b (int): second number

    Returns:
        int: sum of a and b
    """
    return a + b

add(2, 3)
```
