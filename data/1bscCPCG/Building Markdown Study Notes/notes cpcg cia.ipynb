# 🐍 Topic 1: Python Fundamentals

## Introduction to Python

Python is a **high-level**, **interpreted** programming language known for its simplicity and readability. Created by Guido van Rossum in 1991, Python emphasizes code readability with its notable use of significant whitespace. It's widely used in web development, data science, automation, artificial intelligence, and scientific computing.

### ✨ Key Characteristics

- **Easy to Learn**: Python's syntax is clean and intuitive, making it perfect for beginners
- **Interpreted Language**: Code executes line by line, making debugging easier
- **Dynamically Typed**: Variable types are determined at runtime
- **Multi-paradigm**: Supports procedural, object-oriented, and functional programming
- **Extensive Libraries**: Rich ecosystem of packages for virtually any task

---

## 📊 Variables and Data Types

Variables in Python are containers for storing data values. Unlike other languages, Python has no command for declaring a variable – you create one the moment you assign a value to it.

### Primitive Data Types

**Integers (`int`)**: Whole numbers without decimal points
```python
age = 25
year = 2024
```

**Floating Point (`float`)**: Numbers with decimal points
```python
price = 19.99
pi = 3.14159
```

**Strings (`str`)**: Sequence of characters enclosed in quotes
```python
name = "Alice"
message = 'Hello, World!'
```

**Booleans (`bool`)**: True or False values
```python
is_active = True
has_access = False
```

## 🔧 Operators

Python supports various operators for performing operations on variables and values.

### Arithmetic Operators
- `+` Addition
- `-` Subtraction
- `*` Multiplication
- `/` Division
- `//` Floor Division
- `%` Modulus
- `**` Exponentiation

### Comparison Operators
- `==` Equal to
- `!=` Not equal to
- `>` Greater than
- `<` Less than
- `>=` Greater than or equal to
- `<=` Less than or equal to

### Logical Operators
- `and` Returns True if both statements are true
- `or` Returns True if one statement is true
- `not` Reverses the result

---

## 🔀 Control Flow

Control flow statements determine the order in which code executes based on certain conditions.

### Conditional Statements (if-elif-else)

```python
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
```

### Loops

**For Loops**: Iterate over sequences
```python
for i in range(5):
    print(i)

fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
```

**While Loops**: Execute while a condition is true
```python
count = 0
while count < 5:
    print(count)
    count += 1
```

## 📦 Functions

Functions are reusable blocks of code that perform specific tasks. They help organize code and make it more maintainable.

```python
def greet(name, greeting="Hello"):
    """Function with a default parameter"""
    return f"{greeting}, {name}!"

# Function call
message = greet("Alice")
```

### Key Concepts
- **Parameters**: Variables listed in the function definition
- **Arguments**: Values passed to the function when called
- **Return Statement**: Sends result back to the caller
- **Default Parameters**: Provide fallback values
- **Docstrings**: Documentation strings describing function purpose

## 🎯 Python Best Practices

### Code Style (PEP 8)
- Use 4 spaces for indentation
- Keep lines under 79 characters
- Use meaningful variable names (`user_count` not `uc`)
- Add comments for complex logic
- Follow snake_case for variable and function names

### Error Handling

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
finally:
    print("Execution completed")
```

## Introduction to Data Structures

Data structures are specialized formats for organizing, processing, and storing data. Python provides several built-in data structures that are powerful, flexible, and optimized for different use cases. Choosing the right data structure can dramatically improve your program's efficiency and readability. Understanding when and how to use each structure is fundamental to writing effective Python code.

---

## 📝 Lists

Lists are **ordered**, **mutable** collections that can contain items of different data types. They're one of the most versatile data structures in Python.

### Creating Lists

```python
# Empty list
empty_list = []

# List with elements
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
nested = [[1, 2], [3, 4], [5, 6]]
```

### Common List Operations

```python
fruits = ['apple', 'banana', 'cherry']

# Accessing elements
first = fruits[0]        # 'apple'
last = fruits[-1]        # 'cherry'

# Slicing
subset = fruits[0:2]     # ['apple', 'banana']

# Adding elements
fruits.append('date')    # Add to end
fruits.insert(1, 'blueberry')  # Insert at index

# Removing elements
fruits.remove('banana')  # Remove by value
popped = fruits.pop()    # Remove and return last item

# Other operations
length = len(fruits)
fruits.sort()            # Sort in place
fruits.reverse()         # Reverse in place
```

### List Comprehensions

A concise way to create lists based on existing sequences:

```python
# Traditional approach
squares = []
for x in range(10):
    squares.append(x**2)

# List comprehension
squares = [x**2 for x in range(10)]

# With condition
evens = [x for x in range(20) if x % 2 == 0]
```

---

## 🎯 Tuples

Tuples are **ordered**, **immutable** collections. Once created, their elements cannot be changed, making them ideal for data that shouldn't be modified.

```python
# Creating tuples
coordinates = (10, 20)
single_item = (42,)      # Note the comma for single-item tuple
rgb_color = (255, 128, 0)

# Tuple unpacking
x, y = coordinates
r, g, b = rgb_color

# Use cases
# - Function returns multiple values
def get_user_info():
    return ("Alice", 25, "alice@email.com")

name, age, email = get_user_info()
```

**When to use tuples**: Use tuples for data that represents a fixed collection (like coordinates, RGB values, or database records) where immutability provides data integrity.

---

## 🗂️ Dictionaries

Dictionaries store **key-value pairs** and are **unordered** (though insertion order is preserved in Python 3.7+). They provide fast lookups and are incredibly versatile.

### Creating and Using Dictionaries

```python
# Creating dictionaries
student = {
    'name': 'Alice',
    'age': 20,
    'courses': ['Math', 'Physics'],
    'gpa': 3.8
}

# Accessing values
name = student['name']
age = student.get('age', 0)  # Safe access with default

# Adding/updating
student['major'] = 'Computer Science'
student['age'] = 21

# Removing
del student['gpa']
email = student.pop('email', None)

# Iterating
for key, value in student.items():
    print(f"{key}: {value}")

# Dictionary comprehension
squares_dict = {x: x**2 for x in range(5)}
```

### Common Patterns

```python
# Checking key existence
if 'name' in student:
    print(student['name'])

# Merging dictionaries (Python 3.9+)
defaults = {'theme': 'dark', 'notifications': True}
user_prefs = {'theme': 'light'}
settings = defaults | user_prefs
```

---

## 🎲 Sets

Sets are **unordered** collections of **unique** elements. They're perfect for membership testing and eliminating duplicates.

```python
# Creating sets
numbers = {1, 2, 3, 4, 5}
unique_letters = set('hello')  # {'h', 'e', 'l', 'o'}

# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

union = a | b           # {1, 2, 3, 4, 5, 6}
intersection = a & b    # {3, 4}
difference = a - b      # {1, 2}
symmetric_diff = a ^ b  # {1, 2, 5, 6}

# Adding and removing
numbers.add(6)
numbers.remove(1)
numbers.discard(10)  # Doesn't raise error if not found
```

---

## ⚡ Performance Considerations

| Operation | List | Tuple | Dict | Set |
|-----------|------|-------|------|-----|
| Access by index | O(1) | O(1) | - | - |
| Access by key | - | - | O(1) | - |
| Search | O(n) | O(n) | O(1) | O(1) |
| Insertion (end) | O(1) | - | O(1) | O(1) |
| Deletion | O(n) | - | O(1) | O(1) |

---

## 🎓 Choosing the Right Data Structure

**Use Lists when:**
- Order matters
- You need to modify elements
- Duplicates are allowed
- You need index-based access

**Use Tuples when:**
- Data shouldn't change
- You need a hashable collection (for dict keys)
- Representing fixed structures (coordinates, RGB values)

**Use Dictionaries when:**
- You need key-value associations
- Fast lookups are important
- Data has named fields

**Use Sets when:**
- You need unique elements
- Testing membership is frequent
- Mathematical set operations are needed

---

## 💡 Key Takeaways

Mastering Python's core data structures enables you to write efficient, elegant solutions. Lists provide flexible ordered collections perfect for sequences that need modification. Tuples offer immutable alternatives ideal for fixed data. Dictionaries excel at key-value mappings with fast lookups, making them indispensable for structured data. Sets efficiently handle unique collections and set operations. Understanding the performance characteristics and appropriate use cases for each structure allows you to optimize your code. Practice combining these structures – nested dictionaries, lists of tuples, sets of frozen sets – to solve complex problems. The right data structure choice often makes the difference between clunky code and elegant solutions.

## Introduction to Data Visualization in Python

Data visualization transforms raw numbers into visual insights, making complex patterns accessible and understandable. Python offers powerful libraries for creating everything from simple plots to interactive dashboards. Effective visualization is crucial for exploratory data analysis, communicating findings, and making data-driven decisions. The three primary libraries – Matplotlib, Seaborn, and Plotly – each serve different needs, from publication-ready static plots to interactive web-based visualizations.

---

## 📊 Matplotlib: The Foundation

Matplotlib is Python's foundational plotting library, offering fine-grained control over every aspect of your visualizations. While it can be verbose, this control makes it perfect for customized, publication-quality graphics.

### Basic Plotting

```python
import matplotlib.pyplot as plt
import numpy as np

# Simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)
plt.xlabel('X axis', fontsize=12)
plt.ylabel('Y axis', fontsize=12)
plt.title('Sine Wave', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### Common Plot Types

```python
# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()

# Bar chart
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='steelblue')
plt.ylabel('Values')

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='coral', edgecolor='black', alpha=0.7)
```

### Subplots and Layouts

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Line
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Plot')

# Plot 2: Scatter
axes[0, 1].scatter(x, y)
axes[0, 1].set_title('Scatter Plot')

# Plot 3: Bar
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Bar Chart')

# Plot 4: Histogram
axes[1, 1].hist(data, bins=30)
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

## 🎨 Seaborn: Statistical Visualization

Seaborn builds on Matplotlib, providing a high-level interface for statistical graphics with beautiful default styles. It's particularly powerful for visualizing relationships in complex datasets.

### Setup and Styling

```python
import seaborn as sns

# Set style
sns.set_style('whitegrid')
sns.set_palette('husl')

# Load sample dataset
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

### Distribution Plots

```python
# Histogram with KDE
sns.histplot(data=tips, x='total_bill', kde=True, bins=30)

# Box plot
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')

# Violin plot
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
```

### Relationship Plots

```python
# Scatter plot with regression line
sns.regplot(data=tips, x='total_bill', y='tip')

# Pair plot (all variable relationships)
sns.pairplot(iris, hue='species', diag_kind='kde')

# Heatmap (correlation matrix)
correlation = tips.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
```

### Categorical Plots

```python
# Count plot
sns.countplot(data=tips, x='day', hue='sex')

# Bar plot with error bars
sns.barplot(data=tips, x='day', y='total_bill', estimator=np.mean, ci=95)

# Point plot
sns.pointplot(data=tips, x='day', y='total_bill', hue='sex')
```

## 🚀 Plotly: Interactive Visualizations

Plotly creates interactive, web-ready visualizations that allow users to zoom, pan, and hover for details. Perfect for dashboards and exploratory analysis.

### Basic Interactive Plots

```python
import plotly.express as px
import plotly.graph_objects as go

# Interactive scatter plot
df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length',
                 color='species', size='petal_length',
                 hover_data=['petal_width'],
                 title='Iris Dataset')
fig.show()

# Interactive line plot
fig = px.line(x=x, y=y, title='Interactive Sine Wave')
fig.update_traces(line_color='#FF6692', line_width=3)
fig.show()
```

### Advanced Visualizations

```python
# 3D scatter plot
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',
                    color='species', size='petal_length')
fig.show()

# Animated plot
gapminder = px.data.gapminder()
fig = px.scatter(gapminder, x='gdpPercap', y='lifeExp',
                 animation_frame='year', animation_group='country',
                 size='pop', color='continent', hover_name='country',
                 log_x=True, size_max=55, range_x=[100, 100000],
                 range_y=[25, 90])
fig.show()

# Sunburst chart
fig = px.sunburst(df, path=['species', 'sepal_length'],
                  values='petal_length')
fig.show()
```

## 🎯 Best Practices

### Choosing the Right Plot

- **Comparison**: Bar charts, grouped bar charts
- **Distribution**: Histograms, box plots, violin plots
- **Relationship**: Scatter plots, line plots, heatmaps
- **Composition**: Pie charts, stacked bars, treemaps
- **Time series**: Line plots, area charts

### Design Principles

```python
# Good visualization example
plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2.5, color='#2E86AB')
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.title('Temperature Change Over Time', fontsize=14, pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
```

**Key principles:**
- Use clear, descriptive labels and titles
- Choose appropriate color schemes (consider colorblind-friendly palettes)
- Avoid chart junk (unnecessary decorations)
- Ensure proper scaling and aspect ratios
- Add legends when displaying multiple series
- Use consistent styling across related visualizations
