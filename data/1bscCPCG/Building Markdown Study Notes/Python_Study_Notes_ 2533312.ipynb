# Program Design Principles

Program design principles help write clean, efficient, and maintainable code.

## Key Principles
### 1. Modularity
Break programs into small, reusable functions.

### 2. Readability
- Use meaningful variable names
- Proper indentation
- Clear comments

### 3. DRY (Don’t Repeat Yourself)
Avoid writing the same code multiple times.

### 4. Efficiency
Write code that uses less time and memory.

### 5. Scalability
Programs should handle growth in data or users.

### 6. Debugging & Testing
- Test frequently
- Handle errors properly

Example:
```python
def calculate_profit(revenue, cost):
    return revenue - cost
```
