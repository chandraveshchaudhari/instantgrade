## Topic 1: Variables, Data Types, and Operators

### Variables

**Naming Rules:**
- Must start with a letter or underscore
- Can contain letters, numbers, and underscores

### Operators

**Arithmetic Operators:**
- `+` Addition
- `-` Subtraction
- `*` Multiplication

## Topic 2: Why Object-Oriented Programming (OOP)?

### What is OOP?

Object-Oriented Programming is a programming paradigm based on the concept of "objects" that contain data (attributes) and code (methods). OOP organizes code into reusable, modular structures.

### Why Use OOP?

**1. Code Reusability**
- Write code once and reuse it multiple times through classes and inheritance
- Reduces redundancy and development time

**2. Modularity**
- Code is organized into self-contained objects
- Easier to understand, maintain, and debug
- Changes in one part don't affect other parts

**3. Data Encapsulation**
- Hide internal details and expose only necessary information
- Protect data from unintended modifications
- Control access through methods

**4. Real-World Modeling**
- Objects represent real-world entities naturally
- Makes code more intuitive and easier to conceptualize

**5. Maintainability**
- Easier to update and modify code
- Clear structure makes collaboration easier

**6. Scalability**
- Easy to add new features through inheritance and polymorphism
- Supports large-scale application development

### Core OOP Concepts

**1. Classes and Objects**
- **Class**: Blueprint for creating objects (like a template)
- **Object**: Instance of a class (actual implementation)

````python
# Example: Class and Object
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    
    def bark(self):
        return f"{self.name} says Woof!"

# Create objects
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Max", "German Shepherd")

print(dog1.bark())
print(f"{dog2.name} is a {dog2.breed}")
````

**2. Encapsulation**
- Bundling data and methods that work on that data
- Restricting direct access to some components

````python
# Example: Encapsulation with private attributes
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Invalid amount"
    
    def get_balance(self):
        return self.__balance

account = BankAccount("Alice", 1000)
print(account.deposit(500))
print(f"Current balance: ${account.get_balance()}")
# print(account.__balance)  # This would raise an error
````

**3. Inheritance**
- Create new classes based on existing classes
- Child class inherits attributes and methods from parent class

````python
# Example: Inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

class Cow(Animal):
    def speak(self):
        return f"{self.name} says Moo!"

cat = Cat("Whiskers")
cow = Cow("Bessie")

print(cat.speak())
print(cow.speak())
````

**4. Polymorphism**
- Same method name, different implementations
- Objects of different classes can be treated uniformly

````python
# Example: Polymorphism
def animal_sound(animal):
    print(animal.speak())

# Different objects, same interface
animals = [Cat("Tom"), Cow("Daisy"), Dog("Rex", "Bulldog")]

for animal in animals:
    if hasattr(animal, 'speak'):
        print(animal.speak())
    elif hasattr(animal, 'bark'):
        print(animal.bark())
````

### Procedural vs OOP Example

````python
# Procedural approach
def calculate_area_procedural(length, width):
    return length * width

def calculate_perimeter_procedural(length, width):
    return 2 * (length + width)

rect_length = 5
rect_width = 3
print(f"Procedural - Area: {calculate_area_procedural(rect_length, rect_width)}")
print(f"Procedural - Perimeter: {calculate_perimeter_procedural(rect_length, rect_width)}")

# OOP approach
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(5, 3)
print(f"\nOOP - Area: {rect.area()}")
print(f"OOP - Perimeter: {rect.perimeter()}")
````

## Topic 3: Probability Essentials

### What is Probability?

Probability is the measure of the likelihood that an event will occur. It's expressed as a number between 0 and 1:
- **0** = Impossible event
- **1** = Certain event
- **0.5** = Equally likely to occur or not occur

**Formula:**
```
P(Event) = Number of favorable outcomes / Total number of possible outcomes
```

### Basic Probability Concepts

**1. Sample Space (S)**
- Set of all possible outcomes
- Example: Rolling a die → S = {1, 2, 3, 4, 5, 6}

**2. Event (E)**
- Subset of the sample space
- Example: Rolling an even number → E = {2, 4, 6}

**3. Complement**
- Event NOT occurring
- P(not E) = 1 - P(E)

````python
# Simple probability calculation
import random

# Rolling a fair die
sample_space = [1, 2, 3, 4, 5, 6]
favorable_outcomes = [2, 4, 6]  # Even numbers

probability = len(favorable_outcomes) / len(sample_space)
print(f"Probability of rolling an even number: {probability}")
print(f"Probability as percentage: {probability * 100}%")

# Simulate 1000 dice rolls
rolls = [random.choice(sample_space) for _ in range(1000)]
even_count = sum(1 for roll in rolls if roll in favorable_outcomes)
experimental_probability = even_count / 1000
print(f"\nExperimental probability (1000 rolls): {experimental_probability:.3f}")
````

### Probability Rules

**1. Addition Rule (OR)**
- For mutually exclusive events: P(A or B) = P(A) + P(B)
- For non-mutually exclusive events: P(A or B) = P(A) + P(B) - P(A and B)

**2. Multiplication Rule (AND)**
- For independent events: P(A and B) = P(A) × P(B)
- For dependent events: P(A and B) = P(A) × P(B|A)

**3. Conditional Probability**
- P(A|B) = P(A and B) / P(B)
- Probability of A given that B has occurred

````python
# Addition Rule Example: Drawing a card
# Probability of drawing a King OR a Heart

total_cards = 52
kings = 4
hearts = 13
king_of_hearts = 1  # Overlap

p_king = kings / total_cards
p_heart = hearts / total_cards
p_king_and_heart = king_of_hearts / total_cards

p_king_or_heart = p_king + p_heart - p_king_and_heart

print(f"P(King): {p_king:.4f}")
print(f"P(Heart): {p_heart:.4f}")
print(f"P(King OR Heart): {p_king_or_heart:.4f}")
````

````python
# Multiplication Rule Example: Independent events
# Probability of flipping two heads in a row

p_heads = 0.5
p_two_heads = p_heads * p_heads

print(f"P(Heads on first flip): {p_heads}")
print(f"P(Heads on second flip): {p_heads}")
print(f"P(Two heads in a row): {p_two_heads}")

# Simulation
def flip_coin():
    return random.choice(['H', 'T'])

simulations = 10000
two_heads_count = sum(1 for _ in range(simulations) if flip_coin() == 'H' and flip_coin() == 'H')
experimental_prob = two_heads_count / simulations

print(f"\nExperimental probability (10000 trials): {experimental_prob:.4f}")
````

### Common Probability Distributions

**1. Uniform Distribution**
- All outcomes are equally likely
- Example: Fair die, fair coin

**2. Binomial Distribution**
- Number of successes in n independent trials
- Each trial has only two outcomes (success/failure)

**3. Normal Distribution**
- Bell-shaped curve
- Continuous probability distribution
- Many natural phenomena follow this

````python
# Example: Binomial distribution
# Probability of getting exactly k heads in n coin flips
from math import comb

def binomial_probability(n, k, p):
    """
    n: number of trials
    k: number of successes
    p: probability of success on each trial
    """
    return comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Probability of getting exactly 3 heads in 5 coin flips
n_flips = 5
k_heads = 3
p_heads_single = 0.5

prob = binomial_probability(n_flips, k_heads, p_heads_single)
print(f"Probability of exactly {k_heads} heads in {n_flips} flips: {prob:.4f}")

# Calculate for all possible outcomes
print("\nProbability distribution for 5 coin flips:")
for k in range(n_flips + 1):
    prob = binomial_probability(n_flips, k, p_heads_single)
    print(f"P({k} heads) = {prob:.4f}")
````

### Expected Value

Expected value (or mean) is the average outcome you'd expect over many trials.

**Formula:**
```
E(X) = Σ [x * P(x)]
```
where x is each possible outcome and P(x) is its probability

````python
# Expected value example: Rolling a die
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6

expected_value = sum(outcome * prob for outcome, prob in zip(outcomes, probabilities))
print(f"Expected value of a die roll: {expected_value:.2f}")

# Simulation to verify
rolls = [random.randint(1, 6) for _ in range(100000)]
average_roll = sum(rolls) / len(rolls)
print(f"Average from 100,000 rolls: {average_roll:.2f}")
````

### Practical Application: Monte Carlo Simulation

Monte Carlo simulations use random sampling to solve problems that might be deterministic in principle.

````python
# Estimating Pi using Monte Carlo method
def estimate_pi(num_points):
    inside_circle = 0
    
    for _ in range(num_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        # Check if point is inside unit circle
        if x**2 + y**2 <= 1:
            inside_circle += 1
    
    # Ratio of areas: circle/square = π/4
    pi_estimate = 4 * inside_circle / num_points
    return pi_estimate

# Run simulation with different sample sizes
sample_sizes = [100, 1000, 10000, 100000]

import math
print(f"Actual value of π: {math.pi:.6f}\n")

for size in sample_sizes:
    estimate = estimate_pi(size)
    error = abs(estimate - math.pi)
    print(f"Estimate with {size:6d} points: {estimate:.6f} (error: {error:.6f})")
````

### Key Probability Formulas Summary

1. **Basic Probability**: P(E) = n(E) / n(S)
2. **Complement Rule**: P(not E) = 1 - P(E)
3. **Addition Rule**: P(A ∪ B) = P(A) + P(B) - P(A ∩ B)
4. **Multiplication Rule**: P(A ∩ B) = P(A) × P(B) [for independent events]
5. **Conditional Probability**: P(A|B) = P(A ∩ B) / P(B)
6. **Expected Value**: E(X) = Σ[x × P(x)]
