# Core Data Structures in Python

Python provides built-in data structures to store, manage, and manipulate data efficiently.

## Lists: Operations, Slicing and Comprehensions

```python
products = ['Laptop', 'Mobile', 'Tablet', 'Smartwatch']
print(products)
print(products[0])
```

### List Operations

```python
products.append('Headphones')
products.remove('Tablet')
print(products)
print(len(products))
```

### List Slicing

```python
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])
print(numbers[::-1])
```

### List Comprehensions

```python
sales = [20000, 45000, 60000, 32000]
high_sales = [s for s in sales if s > 40000]
print(high_sales)
```

## Dictionaries: Key–Value Pairs and Methods

```python
employee = {
    'name': 'Aarna',
    'role': 'Analyst',
    'salary': 55000
}
print(employee)
```

### Dictionary Methods

```python
print(employee.keys())
print(employee.values())
print(employee.get('salary'))
```

## Sets: Unique Elements and Set Operations

```python
customers = {'A', 'B', 'C', 'A', 'B'}
print(customers)
```

### Set Operations

```python
region1 = {'A', 'B', 'C'}
region2 = {'B', 'C', 'D'}
print(region1 | region2)
print(region1 & region2)
print(region1 - region2)
```

## Practice: Dictionaries and Sets

```python
orders = ['P1', 'P2', 'P1', 'P3', 'P2']
unique_orders = set(orders)
print(unique_orders)
```

## Tuples and Other Built-in Types

```python
coordinates = (12.5, 77.6)
print(coordinates)
```

## Choosing the Right Data Structure for Business Problems

- Lists → ordered data (monthly sales)
- Dictionaries → structured records (employee data)
- Sets → unique values (customer IDs)
- Tuples → fixed data (coordinates, constants)

## Summary
Core data structures are essential for efficient business analytics and ML workflows.
