1. Object Orientated Programming (OOP)

Object Orientated Programming (OOP) is a programming paradigm that organises code around objects rather than functions and logic. Real-world entities are represented by objects, which integrate behaviour (methods) and data (attributes) into a single unit.

Programs that are modular, reusable, scalable, and simple to maintain can be created with the aid of OOP.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

2. Classes and Objects

🔹 Class

A class is a blueprint or template used to create objects.
It defines:

Attributes (variables)

Methods (functions)

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"My name is {self.name} and I am {self.age} years old."
```

🔹 Object

An object is an instance of a class.
One class can have many objects.
Objects store real data.

```python
s1 = Student("Shrinjini", 19)
print(s1.introduce())
```

5. Inheritance and Polymorphism

🔹 Inheritance

Inheritance allows a class to reuse properties of another class.
In the example given below the student is inheriting from the person.

```python
class Person:
    def speak(self):
        return "I can speak"

class Student(Person):
    def study(self):
        return "I am studying"
```

```python
s = Student()
print(s.speak())
print(s.study())
```

🔹 Polymorphism

Polymorphism means same method name, different behavior.

```python
class Dog:
    def sound(self):
        return "Bark"

class Cat:
    def sound(self):
        return "Meow"
```

```python
animals = [Dog(), Cat()]
for a in animals:
    print(a.sound())
```

6. Types of Methods

🔹 Instance Methods

Work with object data

Use self

```python
class Car:
    def drive(self):
        return "Car is moving"
```

🔹 Class Methods

Work with class data

Use @classmethod and cls

```python
class College:
    name = "ABC College"

    @classmethod
    def get_name(cls):
        return cls.name
```

🔹 Static Methods

Independent of class & object

Use @staticmethod

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b
```

7. Decorators: Enhancing Functions and Methods

Decorators modify the behavior of a function without changing its code.

```python
def my_decorator(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

8. Interactions Between Classes

Classes can interact through:

Composition (has-a relationship)

Association

```python
class Engine:
    def start(self):
        return "Engine started"

class Car:
    def __init__(self):
        self.engine = Engine()

    def drive(self):
        return self.engine.start()
```
