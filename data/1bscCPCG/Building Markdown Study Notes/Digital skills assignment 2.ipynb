## 1. Writing Your First Python Program

### What is Python?
Python is a high-level, interpreted programming language known for its simplicity and readability.

### Features of Python
- Easy to learn and use
- Interpreted language
- Extensive libraries
- Platform independent

### Your First Python Program
```python
print("Hello, World!")
```

### Explanation
- `print()` is a built-in function used to display output
- Text inside quotes is called a **string**
- Python is case-sensitive

## 2. Program Design Principles

### What is Program Design?
Program design is the process of planning a program before writing code.

### Key Design Principles
- **Clarity**: Code should be readable and understandable
- **Modularity**: Break code into functions or blocks
- **Efficiency**: Use optimal logic and resources
- **Reusability**: Avoid repetition

### Example
```python
def add(a, b):
    return a + b

print(add(5, 3))
```

### Benefits of Good Design
- Easier debugging
- Better collaboration
- Improved performance

## 3. Math & Notation Foundations

### Numbers in Python
- **Integer (int)**: Whole numbers → `5`, `-2`
- **Float (float)**: Decimal numbers → `3.14`, `2.5`

### Mathematical Operators
| Operator | Meaning |
|--------|--------|
| `+` | Addition |
| `-` | Subtraction |
| `*` | Multiplication |
| `/` | Division |
| `//` | Floor Division |
| `%` | Modulus |

### Example
```python
x = 10
y = 3

print(x + y)
print(x / y)
print(x % y)
```

### Mathematical Notation in Programming
- Variables represent values
- Expressions follow **PEMDAS**
- Assignment uses `=`
