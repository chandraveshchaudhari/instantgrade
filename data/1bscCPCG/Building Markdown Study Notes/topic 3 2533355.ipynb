## 1. List Comprehensions & Generator Expressions

**Why use them?**
- Cleaner code
- Faster execution
- More Pythonic

**Difference:**
- List comprehension stores data in memory
- Generator produces values one at a time

```python
# List comprehension
squares = [x**2 for x in range(1, 6)]
print(squares)

# Generator expression
gen_squares = (x**2 for x in range(1, 6))
for val in gen_squares:
    print(val)
```

## 3. Error Handling and Exceptions

**Why important?**
- Prevent program crashes
- Handle user input errors
- Make programs robust

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
finally:
    print("Execution complete")
```
