6️⃣ Fundamentals of Hypothesis Testing

A statistical technique for using data to make decisions is hypothesis testing.

🔹 Important Terms

Null Hypothesis (H₀) → No difference or effect

Alternative Hypothesis (H₁) → An impact

Level of Significance (α) → Typically 0.05

p-value → The likelihood that results were observed by chance

🔹 Rule of Decision

Reject H if p-value ≤ α →

Fail to reject H if p-value > α →

🔹 Conceptual Example

H0: Grades are unaffected by average study time

H₁: Grades are impacted by average study time

📌 Frequently employed in A/B testing and business decisions.
