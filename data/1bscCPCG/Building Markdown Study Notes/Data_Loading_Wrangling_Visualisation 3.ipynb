## 3.1 Data Loading (CSV, Excel, SQL, APIs)

**Data Loading:** The process of importing data from external sources into a Python environment for analysis.

**Common Data Sources:**
- CSV files
- Excel spreadsheets
- SQL databases
- Web APIs

```python
# Loading data from a CSV file
import pandas as pd

data_csv = pd.read_csv('sample.csv')
data_csv.head()
```

```python
# Loading data from an Excel file
data_excel = pd.read_excel('sample.xlsx')
data_excel.head()
```

## 3.2 Exploratory Data Analysis (EDA)

**Exploratory Data Analysis:** A technique used to understand data patterns, trends, and anomalies before formal modeling.

**Objectives of EDA:**
- Summarize key characteristics
- Identify patterns and outliers
- Support data-driven decisions

```python
# Basic EDA operations
data_csv.info()
data_csv.describe()
```

## 3.3 Data Cleaning & Preprocessing

**Data Cleaning:** The process of correcting or removing inaccurate, incomplete, or irrelevant data.

**Preprocessing Includes:**
- Removing duplicates
- Standardizing formats
- Correcting data types

```python
# Removing duplicate records
data_cleaned = data_csv.drop_duplicates()
data_cleaned.shape
```

## 3.4 Handling Missing Data & Outliers

**Missing Data:** Occurs when no value is stored for a variable.

**Outliers:** Extreme values that deviate significantly from other observations.

```python
# Handling missing values
data_filled = data_cleaned.fillna(method='ffill')

# Detecting outliers using IQR
Q1 = data_filled.quantile(0.25)
Q3 = data_filled.quantile(0.75)
IQR = Q3 - Q1
```

## 3.5 Feature Engineering

**Feature Engineering:** The process of creating new variables from existing data to improve model performance.

**Examples:**
- Encoding categorical variables
- Creating interaction features

```python
# Encoding categorical variables
encoded_data = pd.get_dummies(data_filled, drop_first=True)
encoded_data.head()
```

## 3.6 Data Visualisation

**Data Visualisation:** Graphical representation of data to identify trends and patterns.

```python
# Basic data visualization
import matplotlib.pyplot as plt

data_filled.hist()
plt.show()
```
