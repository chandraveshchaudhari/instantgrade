## 1. Variables, Data Types, and Operators

### Variables
Variables store data values.

```python
x = 10
name = "Gunjan"
price = 99.99

print(x, name, price)
```

### Common Data Types

```python
# Integer
a = 5

# Float
b = 3.14

# String
c = "Hello"

# Boolean
d = True

# List
e = [1, 2, 3]

# Tuple
f = (4, 5, 6)

# Dictionary
g = {"name": "Gunjan", "age": 20}

print(type(a), type(b), type(c), type(d), type(e), type(f), type(g))
```

### Operators

```python
# Arithmetic operators
x = 10
y = 3

print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Modulus:", x % y)

# Comparison
print(x > y)
print(x == y)

# Logical
print(x > 5 and y < 5)
```

## 2. Control Flow (If–Else)

```python
age = 18

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

### If–Elif–Else

```python
marks = 75

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
else:
    grade = "C"

print("Grade:", grade)
```

## 3. Loops (For, While, Nested)

### For Loop

```python
for i in range(5):
    print("Number:", i)
```

### While Loop

```python
count = 0

while count < 5:
    print("Count:", count)
    count += 1
```

### Nested Loop

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

## 4. Functions and Lambda Expressions

### Functions

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Gunjan"))
```

### Function with Default Argument

```python
def power(base, exponent=2):
    return base ** exponent

print(power(5))      # default exponent
print(power(5, 3))   # custom exponent
```

### Lambda Functions

```python
square = lambda x: x * x
print(square(4))

add = lambda a, b: a + b
print(add(3, 5))
```

## 5. Modules and Packages

### Using Built-in Modules

```python
import math

print("Square root:", math.sqrt(16))
print("Pi value:", math.pi)
```

### Import Specific Function

```python
from math import factorial

print("Factorial of 5:", factorial(5))
```

## 6. Best Practices for Business-Oriented Code

### Key Practices
- Use clear and meaningful variable names.
- Write reusable functions.
- Add comments and documentation.
- Handle errors properly.
- Keep code simple and readable.


### Example: Clean Business Function

```python
def calculate_total(price, quantity, tax_rate=0.18):
    """Calculate total bill including tax."""
    subtotal = price * quantity
    tax = subtotal * tax_rate
    total = subtotal + tax
    return total

print(calculate_total(500, 2))
```
