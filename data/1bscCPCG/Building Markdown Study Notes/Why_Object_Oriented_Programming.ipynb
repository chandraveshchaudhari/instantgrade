# Why Object Oriented Programming
---


## Introduction
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of objects. 
Objects contain data (attributes) and functions (methods). OOP helps organize code, improve reusability, and enhance maintainability.

## Why OOP?
### 1. Modularity
Breaks complex programs into smaller, manageable pieces.

### 2. Reusability
Code can be reused using inheritance.

### 3. Encapsulation
Protects data by restricting direct access.

### 4. Abstraction
Hides internal implementation details and shows only essential features.

### 5. Maintainability
Easier to update and debug programs.

## Basic Example of OOP in Python
Below is a simple example of a class and object.

```
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

s1 = Student("Alice", 20)
s1.display()
```
