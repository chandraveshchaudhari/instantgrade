# Python Fundamentals

Python fundamentals form the base for data analysis, automation, and machine learning.

## Variables, Data Types and Operators

Variables are used to store data values in memory.

```python
name = 'Aarna'
age = 18
cgpa = 8.6
is_student = True
print(name, age, cgpa, is_student)
```

### Common Data Types

- int
- float
- str
- bool
- list
- tuple
- dict

```python
numbers = [10, 20, 30]
student = {'name': 'Aarna', 'age': 18}
print(numbers)
print(student)
```

### Operators in Python

```python
x = 10
y = 3
print(x + y)
print(x > y)
print(x > 5 and y < 5)
```

## Control Flow (If-Else)

```python
sales = 45000
if sales > 50000:
    print('High Sales')
else:
    print('Average Sales')
```

## Loops

```python
for i in range(1, 6):
    print('Month', i)
```

```python
count = 1
while count <= 5:
    print('Count:', count)
    count += 1
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

## Functions

```python
def calculate_profit(revenue, cost):
    return revenue - cost

print(calculate_profit(100000, 75000))
```

## Lambda Expressions

```python
profit = lambda revenue, cost: revenue - cost
print(profit(80000, 50000))
```

## Modules and Packages

```python
import math
print(math.sqrt(81))
```

## Best Practices for Business-Oriented Code

- Use meaningful variable names
- Write reusable functions
- Keep code readable
- Avoid hard-coded values
