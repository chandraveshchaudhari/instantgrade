📘 Program Design Principles

Program Design Principles focus on how to plan, structure, write, test, and maintain programs so they are readable, efficient, scalable, and easy to deploy. These principles are especially important in business applications and machine learning systems, where programs grow large and complex over time.

1️⃣ Writing Clean and Modular Code

🔹 Clear Code

Code that's clean is:

Easy to read

Simple to comprehend

Simple to change

✅ Features of Clean Code

Meaningful variable and function names

Brief and targeted functions

Consistent formatting and indentation

Steer clear of needless complexity

```python
# Bad practice
x = 10
y = 5
z = x * y
```

```python
# Clean practice
length = 10
width = 5
area = length * width
```

🔹 Modular Code

Modular programming divides a program into independent, reusable modules.

Benefits:

Easier debugging

Better reusability

Improved collaboration

```python
def calculate_area(length, width):
    return length * width

def calculate_perimeter(length, width):
    return 2 * (length + width)
```

2️⃣ Design Patterns for ML Applications

🔹 What are Design Patterns?

Design patterns are proven solutions to common software design problems.

They offer a common method for organising code.

🔹 Common Design Patterns Used in ML

🧱 Pipeline Pattern

Used to structure ML workflows.

```python
class Pipeline:
    def run(self):
        self.load_data()
        self.preprocess()
        self.train_model()
```

🏭 Factory Pattern

Used to create objects without specifying the exact class.

```python
class ModelFactory:
    def get_model(self, model_type):
        if model_type == "linear":
            return LinearModel()
        elif model_type == "tree":
            return TreeModel()
```

Observer Pattern

Used when one component needs to be notified of changes.

 Example:

Model training progress updates

Logging and monitoring systems
