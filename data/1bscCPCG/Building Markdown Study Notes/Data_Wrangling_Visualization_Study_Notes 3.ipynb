## Data Loading (CSV, Excel, SQL, APIs)
Common ways to load business data into Python.

```python
import pandas as pd

# Load CSV
df_csv = pd.read_csv('data.csv')

# Load Excel
# df_excel = pd.read_excel('data.xlsx')

df_csv.head()
```

## Exploratory Data Analysis (EDA)
Understanding your data before modeling.

```python
df_csv.info()
df_csv.describe()
```

## Data Cleaning & Preprocessing
Fixing messy real-world data.

```python
# Remove duplicates
df_clean = df_csv.drop_duplicates()

# Rename columns
df_clean.columns = [col.lower() for col in df_clean.columns]

df_clean.head()
```

## Handling Missing Data & Outliers

```python
# Check missing values
df_clean.isna().sum()

# Fill missing values
df_clean_filled = df_clean.fillna(df_clean.mean(numeric_only=True))

df_clean_filled.head()
```

## Feature Engineering
Creating new useful features.

```python
# Example feature
df_clean_filled['log_sales'] = None  # placeholder example

df_clean_filled.head()
```

## Data Visualisation

```python
import matplotlib.pyplot as plt

plt.figure()
df_clean_filled.hist()
plt.show()
```
